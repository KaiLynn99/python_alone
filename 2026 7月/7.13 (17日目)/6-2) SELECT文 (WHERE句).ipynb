## 3. `WHERE`句

- `WHERE`句は、特定の行に対して比較演算や論理演算を行い、結果が`TRUE`になる行だけをフィルタリングするために使用する。

```sql
SELECT カラム名
FROM テーブル名
WHERE カラム値 比較演算子 / 論理演算子;
```

#### 1) 比較演算子

- `比較演算子`は、数値・文字列・日付などに使用できるが、多くの場合は`数値データのカラム値を比較する`際に使用される。<br> (<, <=, =, >, >=, <>, !=)

````python
%%sql

SELECT film_id, title, length
FROM film
WHERE length <= 50
LIMIT 5;
````

````python
%%sql

SELECT film_id, title, length
FROM film
WHERE length = 50;
````

````python
%%sql

SELECT film_id, title, length
FROM film
WHERE length <> 50 # != も可能
LIMIT 5;
````

### 2) 論理演算子

`WHERE`句では、条件を組み合わせるために論理演算子（Logical Operator）を使用する。

| 論理演算子 | 意味 | 例 |
|---|---|---|
| `AND` | すべての条件が真 | `A AND B` |
| `OR` | いずれかの条件が真 | `A OR B` |
| `NOT` | 条件の反対 | `NOT A` |
| `IN` | 複数の値のうち、いずれかに一致 | `A IN (...)` |
| `NOT IN` | 複数の値のいずれにも一致しない | `A NOT IN (...)` |
| `BETWEEN` | 指定した範囲に含まれる | `A BETWEEN 1 AND 10` |
| `NOT BETWEEN` | 指定した範囲に含まれない | `A NOT BETWEEN 1 AND 10` |
| `LIKE` | 文字列パターンに一致 | `A LIKE 'A%'` |
| `NOT LIKE` | 文字列パターンに一致しない | `A NOT LIKE 'A%'` |
| `IS NULL` | `NULL`である | `A IS NULL` |
| `IS NOT NULL` | `NULL`ではない | `A IS NOT NULL` |
| `EXISTS` | サブクエリの結果が存在する | `EXISTS (...)` |

````python
%%sql

SELECT *
FROM customer
WHERE active = 1
AND store_id = 1
LIMIT 3;
````

````python
%%sql OR_result <<

SELECT *
FROM customer
WHERE store_id = 1
OR store_id = 2;
````

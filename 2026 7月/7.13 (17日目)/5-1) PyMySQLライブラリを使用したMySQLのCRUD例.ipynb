## 0. 実習用データベースとテーブルの作成

```python
%%sql

CREATE DATABASE ex;

USE ex;

CREATE TABLE product1(
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(30),
    price INT
);
```

## 1. MySQLに接続

```python
import pymysql

conn = pymysql.connect(
    host="localhost",
    user="root",
    password="class1234",
    database="ex",
    charset="utf8"
)

cursor = conn.cursor()

print("MySQL接続成功")
```

## 2. INSERT

```python
products = [
    {"name": "キーボード", "price": 50000},
    {"name": "マウス", "price": 25000},
    {"name": "モニター", "price": 300000},
    {"name": "ノートパソコン", "price": 1200000},
    {"name": "USB", "price": None}, # PythonではSQLのNULLをNoneで表す。
    {"name": "ウェブカメラ", "price": 80000},
    {"name": "ヘッドセット", "price": 90000},
    {"name": "スピーカー", "price": 70000},
    {"name": "外付けハード", "price": 140000},
    {"name": "プリンター", "price": 220000}
]

sql = """
INSERT INTO product1(name, price)
VALUES(%s, %s)
"""

for product1 in products:
        cursor.execute(sql, (product1["name"], product1["price"]))

conn.commit() # テーブルのデータを変更するINSERT / UPDATE / DELETEは、実行後に必ずcommitしなければ最終的に反映されない

print("10点の商品が追加されました。")
```

## 3. SELECT

```python
sql = "SELECT * FROM product1"

cursor.execute(sql) # sql変数に保存されたSQL文を実行する。

rows = cursor.fetchall() # 実行結果のすべての行を取得する。

for row in rows:
    print(row)
```

## 4. UPDATE

```python
update_price_val = 150000 # 修正する価格を変数に保存する。
update_id = 1 # 修正対象の商品IDを変数に保存する。

sql = """
update product1
set price = %s
where id = %s
"""

cursor.execute(sql, (update_price_val, update_id)) 

conn.commit()

print(f"id {update_id}番の修正が完了しました。")
```

## SELECT

```python
sql = "SELECT * FROM product1"

cursor.execute(sql)  

rows = cursor.fetchall()  

for row in rows:
    print(row)
```

## 5. DELETE

```python
sql = """
DELETE FROM product1
WHERE price is NULL
"""

cursor.execute(sql)
conn.commit()

print(f"{cursor.rowcount}行のデータを削除しました。")
```

## SELECT

```python
sql = "SELECT * FROM product1"

cursor.execute(sql)

rows = cursor.fetchall()

for row in rows:
    print(row)
```

## 6. MySQL データベース接続終了

```python
try:
    sql = """
    DELETE FROM product1
    WHERE price is NULL
    """
    cursor.execute(sql)
    print(f"{cursor.rowcount}行のデータを削除しました。")
    conn.commit()
except:
    print("DB例外が発生しました。")
finally:
    cursor.close() 
    conn.close()
```

#### ※ `try-except-finally`は、DB処理中に例外が発生した場合でも、`最後に必ずDB接続を終了する`ために使用する。 

- DB接続を開いたままにすると、不要な接続が残り、DBリソースを無駄に使用する可能性がある。
- そのため、`cursor.close()`と`conn.close()`は、必ず実行される`finally`に書くのが基本である。
- ただし、Jupyter NotebookでまだDB操作を続ける場合は、途中で接続を終了しない。

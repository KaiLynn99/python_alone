- `SELF JOIN`（自己結合）とは、`1つのテーブルを同じテーブル自身と結合すること`である。

- 主に、同じテーブル内で関連付けられている階層構造のデータを取得する際に使用する。

## 1. `employee`テーブルの作成

- `emp_id`：従業員番号（主キー）
- `emp_name`：従業員名
- `manager_id`：上司の従業員番号

`manager_id`には、同じ`employee`テーブルにある上司の`emp_id`が保存される。

```python
%%sql

CREATE TABLE employee (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(30),
    manager_id INT
);
```

`employee`テーブル内で、`manager_id`が同じテーブルの`emp_id`を参照する。

```python
%%sql

INSERT INTO employee VALUES
(1, 'ハウル', NULL),
(2, '千尋', 1),
(3, 'ソフィー', 2),
(4, 'パズー', 2),
(5, 'シータ', 3),
(6, 'キキ', 4);
```

## 2. SELF LEFT JOIN

`SELF LEFT JOIN`は、同じテーブルに異なる別名を付け、テーブル自身と左外部結合する方法である。

`LEFT JOIN`を使用するため、対応する上司が存在しない従業員も含め、すべての従業員が取得される。

```python
%%sql

SELECT
    e.emp_name AS 従業員,
    m.emp_name AS 上司
FROM employee AS e -- 同じ`employee`テーブルに、それぞれ異なる別名を付けて結合する。
LEFT JOIN employee AS m
    ON e.manager_id = m.emp_id;
```

## 3. SELF INNER JOIN

`SELF INNER JOIN`は、同じテーブルに異なる別名を付け、テーブル自身と内部結合する方法である。

`INNER JOIN`を使用するため、対応する上司が存在する従業員のみが取得される。

```python
%%sql

SELECT
    e.emp_name AS 従業員,
    m.emp_name AS 上司
FROM employee AS e
INNER JOIN employee AS m
    ON e.manager_id = m.emp_id;
```

#### # SELF LEFT JOINとの違い

| 結合方法 | 上司がいない従業員 |
|---|---|
| `SELF LEFT JOIN` | `NULL`として表示される |
| `SELF INNER JOIN` | 結果から除外される |

## 4. SELF CROSS JOIN

`SELF CROSS JOIN`は、同じテーブルに異なる別名を付け、テーブル自身と`CROSS JOIN`する方法である。

同じテーブル内のデータ同士で、考えられるすべての組み合わせを作成する際に使用する。

```python
%%sql

SELECT
    s1.student_name AS 学生1,
    s2.student_name AS 学生2
FROM student AS s1
CROSS JOIN student AS s2
WHERE s1.student_name <> s2.student_name;
```

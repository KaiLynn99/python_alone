## 1. リスト (List)

- 他のオブジェクトを集めてグループ化できる整列されたコレクションである。

- リストに含まれるオブジェクトアイテムはそれぞれ独自の位置を持つ。 インデックス作成やスライス作業を行うことができる。

- 任意のオブジェクトを含めることができる。 リストの中にリストオブジェクトを持つネストも可能である。

- 可変タイプで、サイズを大きくしたり小さくしたりできる。 

- 文字列（不変）がサポートしない削除やインデックス割り当て演算など、リストを直接変更する演算もサポートする。

## 2. リストの演算

#### 1) インデックス作成

```python
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# インデックス割り当て演算
nums[2] = 99 # リストは可変オブジェクト
print(nums)
```

#### 2) スライシング：スライシング演算はリスト自体が変更されるのではなく、スライスされたリストを返す。

```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) # [10, 20, 30, 40, 50] スライス演算後でもリスト自体は変更されない。
```

## 3. リストメソッド

#### - append() : 最後に1つ追加

```python
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

#### - insert() : 希望の位置に追加

```python
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

#### - extend() : 複数追加

```python
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

#### - remove() : 値で削除

```python
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)
```

#### - pop() : インデックスで削除して返却

```python
nums = [10, 20, 30]
result = nums.pop() # 最後のインデックスを削除して返す
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # インデックス1を削除して返却
print(result)
print(nums)
```

#### - del文：指定したインデックスの要素を削除する

```python
nums = [10, 20, 30, 40]
del nums[1]
print(nums)
```

#### - clear() : 全て削除

```python
nums = [10, 20, 30]
nums.clear()
print(nums)
```

#### - sort()：リストの要素を昇順・降順に並べ替える

```python
nums = [50, 10, 40, 20]
nums.sort() # 昇順
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 降順
print(nums)
```

#### - reverse() : 順序を逆にする、並べ替えに関係なく現在の順序を逆にする。

```python
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)
```

#### - count() : 特定の値がいくつあるか

```python
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3
```

#### - index() : 特定の値のインデックス

```python
fruits = ["apple", "banana", "orange"]
print(fruits. index("banana")) # 1
```

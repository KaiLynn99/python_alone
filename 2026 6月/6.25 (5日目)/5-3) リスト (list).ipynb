# 1. リスト

​

- 他のオブジェクトを集めてグループ化できる整列されたコレクションである。

- リストに含まれるオブジェクトアイテムはそれぞれ独自の位置を持つ。 インデックス作成やスライス作業を行うことができる。

- 任意のオブジェクトを含めることができる。 リストの中にリストオブジェクトを持つネストも可能である。

- 可変タイプで、サイズを大きくしたり小さくしたりできる。 

- 文字列（不変）がサポートしない削除やインデックス割り当て演算など、リストを直接変更する演算もサポートする。

```python
# 2. リストの演算


# 1) インデックス作成

nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# インデックス割り当て演算
nums[2] = 99 # リストは可変オブジェクト
print(nums)
```

```python
# 2) スライシング：スライシング演算はリスト自体が変更されるのではなく、スライスされたリストを返す。

nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) # [10, 20, 30, 40, 50] スライス演算後でもリスト自体は変更されない。
```

```python
# 3. リストメソッド


# append() : 最後に1つ追加

fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 希望の位置に追加

fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
# extend() : 複数追加

nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

```python
# remove() : 値で削除

fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)
```

```python
# pop() : インデックスで削除して返却

nums = [10, 20, 30]
result = nums.pop() # 最後のインデックスを削除して返す
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # インデックス1を削除して返却
print(result)
print(nums)
```

```python
# del 文

nums = [10, 20, 30, 40]
del nums[1]
print(nums)
```

```python
# clear() : 全て削除

nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# ソートメソッド

nums = [50, 10, 40, 20]
nums.sort() # 昇順
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 降順
print(nums)
```

```python
# reverse() : 順序を逆にする、並べ替えに関係なく現在の順序を逆にする。

nums = [10, 20, 30, 40]
nums.reverse()
print(nums)
```

```python
# count() : 特定の値がいくつあるか

nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3
```

```python
# index() : 特定の値のインデックス

fruits = ["apple", "banana", "orange"]
print(fruits. index("banana")) # 1
```

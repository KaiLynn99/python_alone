## 3. セット (Set)

- 重複を許容しないデータ型、順序なし（インデックス不可）
- 他の数学オブジェクトとは異なり、重複を持たない特性があるため、基本データ型として多く使用されている。

```python
# 1. セットの集合演算

a = {1, 2, 3}
b = {3, 4, 5}

# 数学用の集合演算が可能である
print(a & b) # 共通部分
print(a | b) # 和集合
print(a - b) # 差集合
```

```python
# 2. 組み込み関数

s = {1,2,3}

s.add(4) # 一つ追加
print(s)

s.update([5,6]) # 複数追加
print(s)

s.remove(2) # 値2を削除（なければエラー）
print(s)

s.discard(10) # 10がなくてもエラーはない (値段がなければ無視)
print(s)

```

```python
s = {1,2,3}

print(len(s)) # 個数
print(2 in s) # 含まれるかどうか

s.clear() # 全て削除
print(s) # set()
```

```python
# {}は空のsetではなく空のdict型である。 空のセットが必要な場合はset()関数を使用

d = {}
print(d)
print(type(d))

s = set()
print(s)
print(type(s))
```

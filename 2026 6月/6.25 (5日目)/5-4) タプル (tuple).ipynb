# リストに似ていますが、丸括弧()を使用します。（要素1、要素2、....、要素n）

# リストの場合は項目を変更できるが、タプルの項目は変更できない。

```python
t = 1, 3, 5 # パッキング実行 (1, 3, 5)

print(t) 

print(type(t))
```

```python
l = list(t)

print(l, type(l))
```

```python
x, y, z = t # アンパッキング実行

print(x, y, z)
```

```python
a, b = [3, 6] # リストの二つの要素をaとbにアンパッキング

print(a, b)
```

```python
# パッキング、アンパッキング時 * 演算子

# a, b = [1, 2, 3, 4] # too many values to unpack
*a, b = [1, 2, 3, 4]

print(a)
print(b)
```

```python
c, *d = (1, 2, 3, 4)

print(c)
print(d)
```

```python
e, *f, g = (1, 2, 3, 4)
print(e)
print(f)
print(g)
```

```python
h = [1, 2, 3]
print(h)
print(*h) # print(h[0], h[1], h[2])
```

```python
# タプルは不変オブジェクトであるため、生成後に要素を追加・削除・修正することはできない。

# タプルで代表的に使用できるメソッド

data = (1, 2, 2, 3)

print(data.count(2)) # 2の数
print(data.index(3)) # 3の位置
```

# 次のようなメソッドはタプルで使用できない。

append()
remove()
sort()
reverse()

```python
# 修正が必要なとき、

# タプルは必要な時、リストに変換してから修正し、再びタプルに戻す。

data = (1, 2, 3)

temp = list(data) # タプル → リスト
temp.append(4) # リストから修正
data = tuple(temp) # リスト → タプル

print(data)
```

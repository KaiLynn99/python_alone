## 2. タプル (Tuple)

- リストに似ているが、丸括弧()を使用する。（要素1、要素2、....、要素n）

- リストの場合は項目を変更できるが、タプルの項目は変更できない。

#### ※ タプルのパッキング（Packing）、アンパッキング（Unpacking）

```python
t = 1, 3, 5 # パッキング実行 (1, 3, 5)

print(t) 

print(type(t))
```

```python
l = list(t)

print(l, type(l))
```

```python
x, y, z = t # アンパッキング実行

print(x, y, z)
```

```python
a, b = [3, 6] # リストの二つの要素をaとbにアンパッキング

print(a, b)
```

#### - パッキング、アンパッキング時 * 演算子

```python
# a, b = [1, 2, 3, 4] # too many values to unpack
*a, b = [1, 2, 3, 4]

print(a)
print(b)
```

```python
c, *d = (1, 2, 3, 4)

print(c)
print(d)
```

```python
e, *f, g = (1, 2, 3, 4)
print(e)
print(f)
print(g)
```

```python
h = [1, 2, 3]
print(h)
print(*h) # print(h[0], h[1], h[2])
```

## 2. リストとタプルの間のデータ型の変更

```python
x = (1,3)
print(x)

y = list(x)
print(y)

z = tuple(y)
print(z)
```

## 3. タプル演算

```python
a = (1,2)
b = (3,4)

print(a + b)    
```

```python
print((1,2) *3)
```

```python
t = (1,2,3)

print(2 in t)
print(5 in t)
```

```python
t = (10,20,30,40)

print(t[1])
print(t[1:3])
```

## 4. タプルメソッドと関連する組み込み関数 

```python
t = (1,2,2,3)

print(t.count(2))   # 2の個数 -> 2
print(t.index(2))   # 2の最初の位置 -> 1
```

#### ※ タプルは不変オブジェクトであるため、リストに変更した後、機能を適用してから、再びタプルに変更する必要がある。

```python
t = (1,2,3)

lst = list(t)   # タプル -> リスト
lst.append(4)

t2 = tuple(lst) # リスト -> タプル
print(t2)   
```

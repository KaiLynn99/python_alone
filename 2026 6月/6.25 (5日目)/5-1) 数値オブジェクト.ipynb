### Pythonの数値（Number）型は多様である。 入門段階では整数（int）、実数（float）、複素数（complex）だけが主に使用されるが、金融計算や高精度計算のためにDecimalやFractionも提供されている。

### データサイエンスでは、ほとんどが浮動小数点（float）型だけを使用する。

#### 1. 整数 (int)
- Pythonのintは整数（Integer）型であり、他の言語（C、Javaなど）とは異なり、固定サイズ（32ビット、64ビット）ではない。

- つまり、メモリが許す限り非常に大きな整数まで表現できる。

- メモリ許容内で正確な整数計算が可能

```python
a = 10
b = 100000000000000000000000000000000000000000

print(a)
print(b)
```

#### 2. 実数 (float)
- 浮動小数点方式を使用 

- 非常に大きな数と小さな数を表現可能 

- 誤差が生じる可能性がある

- 一般的な数値計算、データサイエンスで使用

```python
print(0.1 + 0.2) # 0.30000000000000004
```

#### 3. 数値型をbool型に変換

- 0はFalse、0以外のすべての数はTrue

- bool()関数はオブジェクトをboolオブジェクトに型を変更する関数

```python
a = -1
b = 0
c = 1
d = 2
e = 0.1

print(bool(a)) # True
print(bool(b)) # False
print(bool(c)) # True
print(bool(d)) # True
print(bool(e)) # True
```

#### 4. 数値型とよく使われる組み込み関数

- メソッド ≠ 関数

- メソッド：オブジェクト内に含まれているそのオブジェクト専用の機能

- 関数：あるオブジェクトに依存しない独立したコード（機能を定義した）オブジェクトである。

- 数値型とよく使われる組み込み関数は、数値型を引数として使用する関数である。

- 数値オブジェクトのメソッドは多くなく、使用頻度も低い。

ex)

#### 1) メソッド

- 1.some_fun() : 1という数値オブジェクトに対してのみ何らかの操作を行う機能（メソッド）

#### 2) 関数

- some_fun(1) : 入力された1という数値オブジェクトを使って何らかの操作を行う機能（関数



** - メソッド vs 関数**

```python
n = 10
print(n.bit_length()) # 10の二進数値(1010)の長さ

# bin()関数は引数として入力された10進数の2進数文字列を返す。
# 二進数文字列の前に binaryを意味する"0b"プレフィックスを付けて返す。
print(bin(n)) #0b1010 "0b" + "1010"

# bit_length()メソッドはnという数字オブジェクトを作成する必要があり、
# bin()関数はnオブジェクトとは無関係な独立した関数オブジェクトである。

print(bin(9)) # nに関係なく独立して使用可能 
```

** - 数字を引数としてよく使う組み込み関数**

```python
# 1) 繰り返し乗算 intで返す
print(pow(2, 3)) 

# 2) 商と余りをtupleで返す
print(divmod(10, 3)) 

# 3) 四捨五入された実数値を返す
print(round(3.14159, 2)) 

# 4) 絶対値を返す（返却タイプは引数と同じタイプ）
print(abs(-10)) 
```

# 1．木構造（Tree）

## 1）木構造の目的

木構造とは、階層的なデータを効率的に探索・検索・管理するためのデータ構造である。

木構造の一種である二分探索木では、データを挿入・削除する際に、適切な位置を探す処理が必要になる。

そのため、挿入や削除ではリストより処理コストが高くなる場合もあるが、データを効率的に探索する目的で使用される。


## ※ ``線形構造`と`木構造`の比較

- `線形構造`：先頭から順番にデータを探索する。

- `木構造`：親子関係をたどり、必要な範囲だけを探索することで、目的のデータを効率的に探すことができる。

ただし、木が片側に偏った状態になると、最悪の場合は線形構造と同じような形になり、探索に時間がかかることがある。

## • 線形構造

````python
nums = [10, 20, 30, 40, 50]

# 40を探す
for n in nums:
    if n == 40:
        print("見つかった")
````

```text
[10] → [20] → [30] → [40] → [50]
                 ↑
        順番に1つずつ探索する
```

## • 木構造（ツリー構造）

```text
# 木構造の例

        30
       /  \
     20    40
    /        \
  10          50

# 40を探す

# 30 → 40（すぐに見つかる）
```

```text
        30
       /  \
     20    40 ← 見つかった
    /        \
  10          50
```

## 2）データ分析における木構造

次は、データ分析で重要な**分類（Classification）**の例である。

````python
# 入力データ
age = 35
income = 6000
credit = 750

# 条件分岐を使用してルールを実装する
if age >= 20:
    if income >= 5000:
        if credit >= 700:
            print("承認")
        else:
            print("拒否")
    else:
        print("拒否")
else:
    print("拒否")
````

```text
ルールが増える
      ↓
コード内の if-else が次第に増える
      ↓
管理が難しくなる
（条件が増えると、人が if-else を一つずつ作成して追加することは、
 ほぼ不可能になる）
      ↓
分類条件を自動的に作成できる仕組みが必要になる
```

## 2．if文を木構造で表現する

### • if文

````python
color = "Red"
size = "Large"

if color == "Red":
    if size == "Large":
        print("Apple")
    else:
        print("Cherry")
elif color == "Purple":
    print("Grape")
else:
    print("Unknown Fruit")
````

### • 木構造（ツリー構造）
ここでは、木構造をデータそのものの保存に使用するというより、**条件分岐による処理の順序を表現・保存するために使用する。**

````python
fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "Apple",
        "no": "Cherry"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "Grape",
        "no": "Unknown Fruit"
    }
}
````

### • 木構造で表した条件分岐

条件分岐の処理の流れを、1つの木構造として表現する。

```text
                    color == "Red" ?
                   /                \
                Yes                  No
                 |                    |
        size == "Large" ?     color == "Purple" ?
           /          \          /             \
        Yes            No      Yes              No
         |              |       |                |
       Apple          Cherry   Grape       Unknown Fruit
```

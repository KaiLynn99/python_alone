# 1．累積和（Prefix Sum）

配列（リスト）の特定の区間の合計を高速に求めるために使用するアルゴリズムである。

# 2．累積和配列を作成しておく

累積和が格納されている最後の要素に、元のリストの次の要素を加算し、その結果を `append()` で累積和配列に追加する。

```python
nums = [10, 20, 30, 40, 50]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n)
    # 0+10->10 append [0, 10]
    # 10+20->30 append [0, 10, 30]
    # 30+30->60 append [0, 10, 30, 60]
    # ...  [0, 10, 30, 60, 100, 150]
print(prefix)
```

# 3．区間和の公式

| 配列 | 内容 |
|---|---|
| 元の配列（A） | `[10, 20, 30, 40, 50]` |
| 累積和配列（P） | `[0, 10, 30, 60, 100, 150]` |

※ 計算しやすくするため、累積和配列の先頭に `0` を追加する。

このように累積和をあらかじめ求めておくと、累積和配列（リスト）を利用して区間和を計算できる。

* 配列Aの第1要素から第5要素まで
（インデックス 0～4）の区間和は？

→ 配列 P[5] の値

* 配列Aの第2要素から第5要素まで
（インデックス 1～4）の区間和は？

→ 配列 P[5] - P[2 - 1] の値

* 配列Aの第3要素から第4要素まで
（インデックス 2～3）の区間和は？

→ 配列 P[4] - P[3 - 1] の値

#### * 区間和の公式<br>
区間和 = P[end] - P[start - 1]

```python
nums = [10, 2, 30, 21, 44, 5, 8, 50, 7]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n)

print(prefix)

# 3番目(値 30)から 7番目(値 8)までの合計
start = 3
end = 7

resulit = prefix[end] - prefix[start - 1]

print(resulit)
```

# 4. 実習

scores = [80, 90, 75, 88, 95, 70, 85]

- 1～3番の生徒の合計点は？（scoresリストのインデックス0～2）

- 2～5番の生徒の合計点は？

- 3～7番の生徒の合計点は？

- 1～7番の生徒の合計点は？

```python
scores = [80, 90, 75, 88, 95, 70, 85]

prefix = [0]

for n in scores:
    prefix.append(prefix[-1] + n)

print("点数の累積和配列：", prefix)

print("1～3番の生徒の合計点 :", prefix[3] - prefix[1-1])
print("2～5番の生徒の合計点 :", prefix[5] - prefix[2-1])
print("3～7番の生徒の合計点 :", prefix[7] - prefix[3-1])
print("1～7番の生徒の合計点 :", prefix[7] - prefix[1-1])
```

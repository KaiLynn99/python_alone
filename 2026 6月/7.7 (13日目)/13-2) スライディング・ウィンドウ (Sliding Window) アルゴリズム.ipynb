# 1．スライディングウィンドウ（Sliding Window）

スライディングウィンドウとは、配列やリスト上で固定サイズの区間（Window）を横に移動させながら、問題を効率的に処理するアルゴリズムである。

`start` と `end` の間隔を一定に保ったまま移動する様子が、窓を横に動かす形に似ているため、スライディングウィンドウと呼ばれる。

例えば、次のような生徒の点数リストがあるとする。

```python
scores = [10, 20, 30, 40, 50, 60]
```

ウィンドウサイズを `3` に設定し、連続する3人分の合計点を求めると、次の区間を順番に確認する。

```text
[10, 20, 30]
    [20, 30, 40]
        [30, 40, 50]
            [40, 50, 60]
```

- 連続する3人の合計点を求める。
- ウィンドウのサイズは固定したまま、右方向へ一つずつ移動する。

# 2．スライディングウィンドウの仕組み

ウィンドウのサイズが `3` の場合、ウィンドウは次のように移動する。

```text
[10, 20, 30] 40 50 60
```

移動前のウィンドウの合計：

```text
10 + 20 + 30 = 60
```

これは、リストの先頭からウィンドウサイズ分までの合計である。

```python
sum(scores[0:3])
```


## ウィンドウを右へ1つ移動する

ウィンドウを1つ移動すると、次のように合計を更新する。

```text
現在の合計 ＋ 新しく入る値 − ウィンドウから外れる値
```

```text
60 + 40 - 10 = 90
```

```text
10 [20, 30, 40] 50 60
```


## もう一度右へ1つ移動する

```text
90 + 50 - 20 = 120
```

```text
10 20 [30, 40, 50] 60
```


## さらに右へ1つ移動する

```text
120 + 60 - 30 = 150
```

```text
10 20 30 [40, 50, 60]
```


> 新しい合計 = 現在の合計 + 新しく入る値 - ウィンドウから外れる値

````python
scores = [10, 20, 30, 40, 50, 60]

k = 3

# 最初のウィンドウ（インデックス0から始まる区間）の合計を初期値として設定する
window = sum(scores[:k])

print(window)

# ウィンドウサイズkの位置から、1つずつ右に移動しながら区間内の合計を更新する
for i in range(k, len(scores)): 
    window = window - scores[i-k] + scores[i]
    print(window)
````

> スライディングウィンドウとは、**固定サイズの区間を1つずつ移動させながら、直前の計算結果を再利用するアルゴリズム**である。

# 3．ウィンドウ区間の平均を求める

````python
scores = [80, 90, 70, 100, 95, 85]

k = 3

window = sum(scores[:k])

print(window / k)

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]
    print(window / k)
````

# 4．ウィンドウ区間の合計の最大値（最大区間和）を求める

````python
scores = [10, 20, 30, 40, 50, 60]

k = 3

window = sum(scores[:k])

max_sum = window

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]

    if window > max_sum:
        max_sum = window

print(max_sum)
````

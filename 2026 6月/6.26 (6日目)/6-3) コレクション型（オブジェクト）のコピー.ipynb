\* ネストされたコレクションのコピー時の注意事項

- オブジェクトの copy() メソッドは、現在のオブジェクトのみをコピーする浅いコピー（Shallow Copy）になる。

- 外側のリストは新しく生成される。

- 内側のリストは copy() を使用せず、代入（共同参照）を行う。

- 内側のリストの参照（アドレス）のみをコピーする。

```python
a = [[1, 2], [3, 4]]
b = a.copy()

# 外部のリストは別のオブジェクト
print(a is b)      # False

# 内部のリストは同じオブジェクト
print(a[0] is b[0]) # True
```

```python
a = [[1, 2], [3, 4]]

b = a.copy()

b[0][0] = 999 # 内部リストは相互参照

print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], [3, 4]]

b[1] = 7 # 外部リストは別のオブジェクト
print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], 7]
```

\* ネストされたオブジェクトまですべてコピーする場合は、copyモジュールのdeepcopy()関数を使用して、ディープコピー（Deep Copy）を行う必要がある。

```python
import copy

a = [[1, 2], [3, 4]]

b = copy.deepcopy(a)

b[0][0] = 999

print(a)
print(b)
```

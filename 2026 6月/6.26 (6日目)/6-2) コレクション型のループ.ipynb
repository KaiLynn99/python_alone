- コレクションオブジェクトは、for文を使用することで、自身が持つアイテムを順に処理することができる。

- 文は値を返さない。

- コロン(:)とインデントを構文として使用し、一緒に実行される文の集まりであるコードブロックを表す。

- 文には式を含めることができる。for文では、in演算式が使用される。


```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
colors = ("red", "green", "blue")

for color in colors:
    print(color)
```

```python
nums = {10, 20, 30}

for num in nums:
    print(num)
```

```python
student = {
    "name": "パク・キョンピル",
    "age": 27
}

# キー
for key in student:
    print(key)

# 値
for value in student.values():
    print(value)

# キーと値
for key, value in student.items():
    print(key, value)
```

## 条件式は式であり、条件文は文である。

1. 条件表現式（三項演算子）

* 条件に応じて1つの値を返す。

```python
age = 27
result = "成人" if age >= 19 else "未成年"
print(result)

print(f"あなたは {'成人' if age >= 19 else '未成年'} です") # ""の中の文字列は''を使おう。
```

2. 条件文（if文）

* 文は値を返さない。
* コロン文字（:）とインデント構文を使用して、一緒に実行される文のまとまりであるコードブロックを表現する。

| コード / 構文 | 説明 |
| :--- | :--- |
| `if age >= 19` | ヘッドライン（構文キーワード） |
| `:` | コードブロックの開始を知らせる構文 |
| インデント | **コードブロックに属する文を表す構文** |

```python
age = 20
if age >= 19:
    print("成人です。")
```

```python
age = 15
if age >= 19:
    print("成人です。")
else:
    print("未成年です。")
```

```python
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
```

```python
age = 20
has_ticket = True
if age >= 19 and has_ticket:
    print("入場可能")
```

```python
is_member = False
coupon = True
if is_member or coupon:
    print("割引可能")
```

```python
is_student = False
if not is_student:
    print("学生ではありません。")
```

※ Pythonでは、以下の値はFalseと判定される。

False
None
0
0.0
''
[]
()
{}
set()

```python
data = []
if data:
    print("データあり")
else:
    print("データなし")
```

* ネスト

```python
age = 20
score = 85

if age >= 19:
    if score >= 80:
        print("合格")
```

```python
num = 7

if num % 2 == 0:
    print("偶数")
else:
    print("奇数")
```

```python
num = -5

if num > 0:
    print("正の数")
elif num < 0:
    print("負の数")
else:
    print("0")
```

```python
scores = [95, 60, 80, 45]

for score in scores:
    if score >= 70:
        print("合格")
    else:
        print("不合格")
```

```python
numbers = [3, 8, 11, 20, 25]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} : 偶数")
    else:
        print(f"{num} : 奇数")
```

```python
fruits = ["apple", "banana", "kiwi", "orange"]

for fruit in fruits:
    if len(fruit) >= 6:
        print(fruit)
```

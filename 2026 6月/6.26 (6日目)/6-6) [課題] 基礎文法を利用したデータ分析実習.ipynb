### 問題2. 外れ値の処理

外れ値は次の基準で修正しなさい。

1,000,000 超過 → 1,000,000
0 未満 → 0

```python
from pprint import pprint

for order in orders:
    if order["購入金額"] is not None:
        if order["購入金額"] > 1000000:
            order["購入金額"] = 1000000
            
        elif order["購入金額"] < 0:
            order["購入金額"] = 0

pprint(orders, sort_dicts=False, width=200)
```

### 問題4. 欠損値の処理

欠損値を除いた購入金額の平均を計算した後、欠損値をその平均値で置き換えなさい。

```python
from pprint import pprint

prices = []

for order in orders:
    if order["購入金額"] is not None:
        prices.append(order["購入金額"])

avg = round(sum(prices) / len(prices))

for order in orders:
    if order["購入金額"] is None:
        order["購入金額"] = avg

pprint(orders, sort_dicts=False)
```

### 問題 8. 地域別の平均購入金額

地域別の平均購入金額を求めなさい。

```python
area_group = {}

for order in orders:
    area = order["地域"]
    
    if area not in area_group:
        area_group[area] = []
    
    area_group[area].append(order["購入金額"])


for area, prices in area_group.items():
    avg = round((sum(prices) / len(prices)))
    print(f"{area}: {avg}")
```

### 問題 9. 会員ランク別の平均購入金額

VIP会員と一般会員の平均購入金額を求めなさい。

```python
rank_group = {}

for order in orders:
    grade = order["会員ランク"]

    if grade not in rank_group:
        rank_group[grade] = []

    rank_group[grade].append(order["購入金額"])

for grade in ["VIP", "一般"]:
    prices = rank_group[grade]
    
    avg = round(sum(prices) / len(prices))
    print(f"{grade}: {avg}")
```

### 問題 10. 加入年数別のグループ分析


加入年数を基準に、次のようにグループ分けしなさい。

- 0～2年

- 3～5年

- 6年以上

各グループの平均購入金額を求めなさい。

```python
period_0_2_sum = 0
period_0_2_count = 0

period_3_5_sum = 0
period_3_5_count = 0

period_6_or_more_sum = 0
period_6_or_more_count = 0

for order in orders:
    if 0 <= order["加入年数"] <= 2:
        period_0_2_sum += order["購入金額"]
        period_0_2_count += 1
    elif 3 <= order["加入年数"] <= 5:
        period_3_5_sum += order["購入金額"]
        period_3_5_count += 1
    else:
        period_6_or_more_sum += order["購入金額"]
        period_6_or_more_count += 1

print("0～2年:", round(period_0_2_sum / period_0_2_count))
print("3～5年:", round(period_3_5_sum / period_3_5_count))
print("6年以上:", round(period_6_or_more_sum / period_6_or_more_count))
```

## データ分析演習

### [データ収集]

#### 1) データの準備

```python
students = [

    {"名前": "荻野千尋", "成績": 95, "住民登録番号": "160405-4125830"},

    {"名前": "ハク", "成績": 88, "住民登録番号": "140115-3264710"},

    {"名前": "ソフィー", "成績": None, "住民登録番号": "080510-4381920"},  # 欠損値

    {"名前": "ハウル", "成績": 77, "住民登録番号": "990304-1748260"},

    {"名前": "アシタカ", "成績": 105, "住民登録番号": "090421-3519470"},  # 外れ値

    {"名前": "サン", "成績": 82, "住民登録番号": "110512-4628150"},

    {"名前": "パズー", "成績": -10, "住民登録番号": "130305-3847260"},  # 外れ値

    {"名前": "シータ", "成績": 91, "住民登録番号": "130615-4916380"},

    {"名前": "キキ", "成績": 85, "住民登録番号": "130622-4279510"},

    {"名前": "月島雫", "成績": None, "住民登録番号": "120514-4562830"},  # 欠損値

    {"名前": "天沢聖司", "成績": 79, "住民登録番号": "110601-3185740"},
]


# print(students) よりも、Jupyter Notebookでは students のみを記述したほうが見やすい。

len(students)
```

### ** [データ前処理] **

#### 2) 欠損値の確認

```python
missing_count = 0

for student in students:
    if student["成績"] is None:
        missing_count += 1
        
print("欠損値の数:", missing_count)
```

#### 3) 外れ値の検出（正常範囲：0 〜 100点）

```python
for student in students:
    score = student["成績"]
    
    if student["成績"] is not None:
        if score < 0 or score > 100:
            print(student)
```

#### 4) 外れ値の処理（どう処理するかは要件によって異なる）

```python
for student in students:
    if student["成績"] is not None:
        if student["成績"] > 100:
            student["成績"] = 100

        elif student["成績"] < 0:
            student["成績"] = 0

students
```

#### 5) 欠損値の処理（平均値での補完：どの値で補完するかは要件によって異なる）

```python
scores = []

for student in students:
    if student["成績"] is not None:
        scores.append(student["成績"])

avg = round(sum(scores) / len(scores))

for student in students:
    if student["成績"] is None:
        student["成績"] = avg

students
```

#### 6) 性別の取得 (特徴量抽出)

```python
for student in students:
    
    # 住民登録番号の後ろの席の最初の数字(性別コード)を抽出
    gender_code = student["住民登録番号"].split("-")[1][0]
    
    if gender_code in ["1", "3"]:
        student["性別"] = "男"
    else:
        student["性別"] = "女"

students
```

#### 7) 年齢の計算 (特徴量抽出)

```python
from datetime import datetime

current_year = datetime.now().year


for student in students:
    rrn = student["住民登録番号"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["年齢"] = current_year - birth_year

students
```

### ** [EDA : 探索的データ解析] **

#### 8) 記述統計

```python
scores = []

for student in students:
    scores.append(student["成績"])

print("個数:", len(scores))
print("合計:", sum(scores))
print("平均:", round(sum(scores) / len(scores)))
print("最大:", max(scores))
print("最小:", min(scores))
```

#### 9) 成績別平均成績 (性別ごとの平均成績)

```python
male_sum = 0
male_count = 0

female_sum = 0
female_count = 0

for student in students:
    if student["性別"] == "男":
        male_sum += student["成績"]
        male_count += 1
    else:
        female_sum += student["成績"]
        female_count += 1

print("男 平均:", male_sum / male_count)
print("女 平均:", female_sum / female_count)
```

#### 10) 年代別の平均成績

```python
age_group = {}

for student in students:
    decade = (student["年齢"] // 10) * 10

    if decade not in age_group:
        age_group[decade] = []

    age_group[decade].append(student["成績"])

print(age_group)
```

```python
for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}代 平均: {avg:.2f}")
```

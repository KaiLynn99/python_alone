## 4. 辞書 （Dict）

- テーブルやデータフレームなどの2次元データの1行を表す際によく使用される。

- key : value 構造

- 順序 O (Python 3.7 以上)

- key は重複不可、value は重複可能

1. dict型の生成, 扱い

```python
# 空の辞書

d1 = {}
d2 = dict()

print(d1)
print(d2)
```

```python
# {} 使用

student = {
    "name": "パク・キョンピル",
    "age": 27,
    "score": 99
}

print(student)
print(type(student))
```

```python
# dict() の使用

person = dict(name="パク・キョンピル", age=27, city="ヨンインし")

print(person)
```

```python
# [()]をdictに変換

data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

fruit = dict(data)

print(fruit)
```

\* 辞書（Dictionary）のキー（Key）は必ずハッシュ可能（Hashable）でなければならない。そのため、不変オブジェクトはキーとして使用できるが、可変オブジェクトはキーとして使用できない。

(辞書はキーのハッシュ値を使って、データの保存場所を高速に特定する)
(キーのハッシュ値が途中で変わると、保存したデータを正しく検索できなくなるため、キーはハッシュ可能でなければならない)

```python
d = {"name": "パク・キョンピル"}
print(d)

d = {100: "学生"}
print(d)

d = {
    (1, 2): "座標"
}
print(d)

d = {
    [1, 2]: "リスト"
} # error
print(d)
```

2. dictオブジェクトの演算、メソッド、関数

1) 演算

```python
student = {"name": "パク・キョンピル"}
print(student)

# 要素照会
print(student["name"])

# 追加
student["age"] = 27
print(student)

# 修正
student["age"] = 30
print(student)

# 削除
del student["age"]
print(student)

# in 演算
print("name" in student)
print("age" in student)
```

2) メソッド

```python
student = {"name": "パク・キョンピル"}

print(student.get("name"))
print(student.get("age")) # なければ None を返す

student = {"name": "パク・キョンピル", "age": 27}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 99})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```

3. 関数

```python
student = {
    "name": "パク・キョンピル",
    "age": 27,
    "score": 99
}
print(len(student))

s = {
    "c": 3,
    "a": 1,
    "b": 2
}

print(sorted(s)) # 戻り値はソートされたキーのリスト
print(s) # 元のデータは変更されない

print(list(s)) # 戻り値はキーのリスト
```

# 1. クラス

> クラスとは、`ユーザー定義のデータ型`である。

また、クラスは`オブジェクトを生成するための設計図（ひな型）`であり、オブジェクト指向プログラミング（OOP）の中心となる概念である。

- 関連するデータを`属性`として、データを操作する処理を`メソッド`として、ひとまとまりに定義できる。
- クラスをもとに生成されたオブジェクトを、そのクラスの`インスタンス`という。

# 2. クラスの基本構造

- `class`：クラスを定義するためのキーワード
- `__init__`：コンストラクタ。オブジェクトの生成時に自動的に実行される
- `self`：生成されたオブジェクト自身を表す。Javaの`this`に相当する
- `def`：メソッドを定義するためのキーワード

```py
class クラス名:
    def __init__(self, 引数):
        self.属性 = 値

    def メソッド名(self):
        実行する処理

#### * `メソッド（メンバー関数`）VS `関数`

- `メソッド`は、特定のクラスやインスタンスに属し、その属性や状態を使用・変更する処理である。クラスの内部に定義する。

- `関数`は、特定のクラスやデータ型に依存しない汎用的な処理である。通常はクラスの外部に定義する。

````python
class Student:

    def __init__(self, name, jpn, eng):
        self.name = name
        self.jpn = jpn
        self.eng = eng

    def total(self):
        return self.jpn + self.eng

    def average(self):
        return self.total() / 2

    def print_info(self):
        print(f"名前：{self.name}")
        print(f"国語：{self.jpn}")
        print(f"英語：{self.eng}")
        print(f"合計点：{self.total()}")
        print(f"平均点：{self.average()}")


s1 = Student("カオナシ", 90, 85)
s2 = Student("ハウル", 100, 95)

s1.print_info()
print()

s2.print_info()
````

````python
class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money

    def withdraw(self, money):
        self.balance -= money

    def print_info(self):
        print(f"口座名義人：{self.owner}")
        print(f"残高：{self.balance:,}円")


acc = Account("ポルコ", 10000)

acc.deposit(5000)
acc.withdraw(3000)

acc.print_info()
````

````python
class Car:

    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10

    def print_info(self):
        print(f"車種：{self.model}")
        print(f"速度：{self.speed}km/h")


car = Car("ネコバス")

car.accelerate()
car.accelerate()
car.brake()

car.print_info()
````

````python
class TV:

    def __init__(self):
        self.power = False
        self.channel = 1

    def on(self):
        self.power = True

    def channel_up(self):
        self.channel += 1

    def print_info(self):
        print(f"電源：{self.power}")
        print(f"チャンネル：{self.channel}")


tv = TV()

tv.on()
tv.channel_up()
tv.channel_up()

tv.print_info()
````

````python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height) * 2


r = Rectangle(10, 5)

print(r.area())
print(r.perimeter())
````

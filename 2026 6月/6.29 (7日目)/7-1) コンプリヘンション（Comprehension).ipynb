#### # Pythonの内包表記 （Comprehension）は、既存のシーケンスやデータ構造の要素を組み合わせ、変換、フィルタリングして、新しいリスト、辞書、集合を生成する構文である。

#### # 内包表記は、「ループ + append()」を1行にまとめた構文です

[式 for 変数 in 反復可能オブジェクト]
[式 for 変数 in 反復可能オブジェクト if 条件]

例
[i\*\*2 for i in range(1,6)]  # 1から5まで反復し、i\*\*2の値をリストに格納する

## 1. リスト内包表記


#### # 一般的なループを利用して既存のデータ構造から新しいリストを生成

```python
numbers = []

for i in range(1, 6):
    numbers.append(i)

print(numbers)
```

```python
# 二乗値の生成

squares = []

for i in range(1, 6):
    squares.append(i ** 2)

print(squares)
```

#### # 内包表記

```python
numbers = [i for i in range(1, 6)]
print(numbers)
```

```python
squares = [i ** 2 for i in range(1, 6)]
print(squares)
```

#### # 条件文の追加が可能

```python
evens = [i for i in range(1, 11) if i % 2 == 0]
print(evens)
```

#### ex)

```python
names = ["kim", "lee", "park"]
result = [name.upper() for name in names]
print(result)
```

```python
students = [
    {"名前": "ハク", "成績": 95},
    {"名前": "キキ", "成績": 88},
    {"名前": "アシタカ", "成績": 77}
]

names = [s["名前"] for s in students]
print(names)

scores = [s["成績"] for s in students]
print(scores)

# 優秀学生の抽出

good_students = [
    s["名前"]
    for s in students
    if s["成績"] >= 90
]
print(good_students)
```

## 2. 辞書内包表記

```python
squares = {
    i: i**2
    for i in range(1, 6)
}

print(squares)
```

```python
# 名前と長さ

names = ["ソフィー", "パズー", "月島雫"]
result = {name: len(name) for name in names}
print(result) # {'ソフィー': 3, 'パズー': 3, '月島雫': 3}
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {a.split("-")[0]: a.split("-")[1] for a in accounts}

print(result)
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {
    user: password for user, password in (account.split("-") for account in accounts)
} # accounts 結果 [(,),(,),..]を再び内包表記

print(result)
```

```python
scores = [95, 80, 67, 100]

result = {
    s: "合格" if s >= 80 else "不合格"
    for s in scores
}

print(result)
```

```python
# キーと値の反転

scores = {
    "ナウシカ": 90,
    "ポルコ": 80,
    "メイ": 100
}

result = {value: key for key, value in scores.items()}

print(result)
```

# 1．正規表現のパターン
 
> 正規表現のパターンとは、**文字列の規則を表す式**である。


## 例：電話番号のパターン

### 1）対象となる文字列

次のような電話番号を探すとする。

```text
010-1234-5678
```


### 2）文字列の規則

この電話番号は、次の規則で構成されている。

```text
「010-」＋ 数字4桁 ＋「-」＋ 数字4桁
```


### 3）正規表現で表す

この規則を正規表現で表すと、次のようになる。

```text
010-\d{4}-\d{4}
```


> **ポイント**  
> 正規表現のパターンを作成するためには、さまざまな**メタ文字**を理解する必要がある。

# 2．よく使われる正規表現のメタ文字
  
> Pythonで正規表現を使用して文字列からパターンを探す場合は、`re`モジュールを使用する。

## 1）Raw String

> `r"..."`で表すRaw String（raw文字列）とは、文字列の先頭に `r` を付けることで、バックスラッシュ（`\`）をエスケープ文字として解釈せず、そのまま扱う文字列の表記方法である。


### Raw Stringを使用する理由

正規表現の文法では、バックスラッシュ（`\`）を頻繁に使用する。

しかし、通常の文字列では、そのたびにバックスラッシュを `\\` と記述する必要があり不便なため、Raw Stringを使用することが多い。

### 通常の文字列とRaw Stringの比較

```python
"\\d"   # 通常の文字列
r"\d"   # Raw String
```

上記の2つは、どちらも同じ文字列 `\d` を表している。

| 表記方法 | 記述 | 実際の文字列 |
|---|---|---|
| 通常の文字列 | `"\\d"` | `\d` |
| Raw String | `r"\d"` | `\d` |

### 通常の文字列

通常の文字列では、`\n`が改行として解釈され、`\\`は1つのバックスラッシュとして出力される。

````python
print("Hello\nworld\\")
````

### Raw String

Raw Stringでは、`\n`や`\\`をエスケープシーケンスとして解釈せず、入力した形のまま文字列として扱われる。

````python
print(r"Hello\nworld\\")
````

> **ポイント**  
> 正規表現では、通常の文字列よりもRaw Stringを使用した方が、バックスラッシュを簡潔に記述できる。

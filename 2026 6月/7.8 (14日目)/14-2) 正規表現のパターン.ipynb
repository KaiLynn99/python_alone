# 1．正規表現のパターン
 
> 正規表現のパターンとは、**文字列の規則を表す式**である。


## 例：電話番号のパターン

### 1）対象となる文字列

次のような電話番号を探すとする。

```text
010-1234-5678
```


### 2）文字列の規則

この電話番号は、次の規則で構成されている。

```text
「010-」＋ 数字4桁 ＋「-」＋ 数字4桁
```


### 3）正規表現で表す

この規則を正規表現で表すと、次のようになる。

```text
010-\d{4}-\d{4}
```

# 2．よく使われる正規表現のメタ文字
  
> Pythonで正規表現を使用して文字列からパターンを探す場合は、`re`モジュールを使用する。

## Raw String

> `r"..."`で表すRaw String (raw文字列) とは、文字列の先頭に `r` を付けることで、バックスラッシュ（`\`）をエスケープ文字として解釈せず、そのまま扱う文字列の表記方法である。


### ※ Raw Stringを使用する理由

- 正規表現の文法では、バックスラッシュ（`\`）を頻繁に使用する。

- しかし、通常の文字列では、そのたびにバックスラッシュを `\\` と記述する必要があり不便なため、Raw Stringを使用することが多い。

### 通常の文字列とRaw Stringの比較

```python
"\\d"   # 通常の文字列
r"\d"   # Raw String
```

上記の2つは、どちらも同じ文字列 `\d` を表している。

| 表記方法 | 記述 | 実際の文字列 |
|---|---|---|
| 通常の文字列 | `"\\d"` | `\d` |
| Raw String | `r"\d"` | `\d` |

### • 通常の文字列

通常の文字列では、`\n`が改行として解釈され、`\\`は1つのバックスラッシュとして出力される。

````python
print("Hello\nworld\\")
````

### • Raw String

Raw Stringでは、`\n`や`\\`をエスケープシーケンスとして解釈せず、入力した形のまま文字列として扱われる。

````python
print(r"Hello\nworld\\")
````

## 2）文字クラス `[]`

> 文字クラス `[]` は、角括弧の中に指定した文字のうち、**いずれか1文字に一致するパターン**である。


### `re.findall()` 関数

`re.findall(正規表現, 文字列)` は、文字列全体から正規表現に一致する部分をすべて探し、リストとして返す。

````python
import re

text = "apple banana cat"

result = re.findall(r"[abc]", text)

print(result)
````

`[abc]` は、文字列の先頭から末尾までを調べ、`a`、`b`、`c` のいずれかに一致した文字を順番に返す。

## 3）文字クラスの範囲指定

> 文字クラス `[]` の中でハイフン（`-`）を使用すると、文字の範囲を表すことができる。


- 数字の0から9まで [0123456789] を省略して [0-9] と表記

- [a-z], [A-Z], ...

````python
import re

text = "A12B34"
result = re.findall(r"[0-9]", text)

print(result)

# 0から9までの数字のいずれか1文字に一致すると、返されるリストに追加される
````

### 英大文字を検索する

````python
import re

text = "Python JAVA AI"

result = re.findall(r"[A-Z]", text)

print(result)
````

## 4）否定文字クラス `[^ ]`

> 文字クラスの先頭にある `^` は、**角括弧内に指定した文字を除外し、それ以外の文字に一致することを表す。**

````python
import re

text = "a7bc123"

result = re.findall(r"[^0-9]", text) # 数字ではない文字1つ

print(result)
````

````python
import re

password = input("パスワード：")
result = []

if re.search(r"[ぁ-んァ-ヶ一-龯々ー]", password):
    result.append("日本語は使用できません")

if not re.search(r"[a-z]", password):
    result.append("英小文字が含まれていません")

if not re.search(r"[A-Z]", password):
    result.append("英大文字が含まれていません")

if not re.search(r"[0-9]", password):
    result.append("数字が含まれていません")

if not re.search(r"[^0-9a-zA-Zぁ-んァ-ヶ一-龯々ー]", password):
    result.append("特殊文字が含まれていません")

if len(result) == 0:
    result.append("使用可能なパスワードです!")


result
````

| パターン | 意味 |
|---|---|
| `ぁ-ん` | ひらがな |
| `ァ-ヶ` | カタカナ |
| `一-龯` | 漢字 |
| `々` | 漢字の繰り返し記号 |
| `ー` | 長音記号 |

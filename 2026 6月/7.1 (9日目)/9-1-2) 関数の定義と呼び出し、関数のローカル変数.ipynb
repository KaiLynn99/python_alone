````python
def greet(name):
    print(f"{name}さん、こんにちは")

greet("ナウシカ")
greet("ハウル")

# 関数内の引数も、関数のコードブロック内でのみ使用可能なローカル変数である。
print(name) # NameError: name 'name' is not defined
````

## 7. デフォルト値のパラメータ

#### 関数を呼び出す際に引数を省略すると、関数の定義時にパラメータとして指定されたデフォルト値が使用される。

- デフォルト値を持つ引数のルール

1\) デフォルト値のない仮引数は、デフォルト値のある仮引数より先に記述する。

````python
def greet(name, age=20):
    print(name, age)

greet("サン")
greet("ポルコ", 30)
````

2\) デフォルトパラメータの後に通常のパラメータを記述してはならない。

<u>デフォルトパラメータは常に最後に記述する。</u>

````python
# エラーの発生 : SyntaxError: parameter without a default follows parameter with a default
def greet(name="サン", age):
    print(name, age)
````

- 2つ以上のデフォルトパラメータ

````python
def calc(a, b=10, c=20):
    print(a, b, c)

calc(1)
calc(1, 2)
calc(1, 2, 3)
````

## 8. 関数呼び出し時に`"不変オブジェクトの引数"`と`"可変オブジェクトの引数"`を使用した場合の違い

1\) 不変オブジェクト（Immutable）を引数として渡す。

- int、float、str、tuple、bool など

````python
def func(x):
    x += 10
    print("関数内:", x, id(x))

a = 10

func(a)

print("関数外:", a, id(a))

# 関数内の x のみを変更し、関数外の a はそのまま
# id() の値が異なる
````

2\) 可変オブジェクト（Mutable）を引数として渡す。

- list、dict、set など

````python
def func(nums):
    nums.append(4)
    print("関数内:",nums, id(nums))

data = [1, 2, 3]

func(data)

print("関数外:", data, id(data))

# id()の値が同じである。
````

#### # 関数内でパラメータに新しいオブジェクトを代入すると、パラメータの参照だけが変更され、元の変数はそのまま残る。

````python
def func(nums):
    print("新しい値を代入する前:", nums, id(nums))
    nums = [100, 200]
    print("新しい値を代入した後:", nums, id(nums))

data = [1, 2, 3]
func(data)
print(data, id(data))
````

- 関数を呼び出す際、引数とパラメータは当初、同じオブジェクトを参照する。

- 不変オブジェクトはオブジェクト内部の値を直接変更できないため、演算を行うと新しいオブジェクトが生成される。

- 可変オブジェクトはオブジェクト内部の要素を直接変更できるため、同じオブジェクトを参照している元のものにも変更が反映される。

- パラメータに新しいオブジェクトを代入しても、元のものには変更が反映されない。

## 9. 戻り値の型ヒント

- 関数を宣言する際、関数名の後に戻り値の型を指定する構文を、戻り値の型ヒント（Return Type Hint）という。

```python
def 関数名(パラメータ) -> 戻り値の型:
    ...
    return 値

````python
def add(a: int, b: int) -> int:
    return a + b

def greet(name: str) -> str:
    return f"こんにちは、 {name}"

def get_numbers() -> list:
    return [1, 2, 3]

def get_student() -> dict:
    return {"name": "カオナシ", "age": 20}

def hello() -> None:
    return None

print(add(10, 5))
print(greet("サン"))
print(get_numbers())
print(get_student())
print(hello())
````

## 10. 関数内でのグローバル変数の使用

- 関数の外で定義された変数をグローバル変数という。
- 関数がグローバル変数に依存すると保守性が下がるため、必要な値は引数として渡す方がよい。

````python
x = 10  # グローバル変数

def use_global():
    print(x)  # グローバル変数を参照

def use_local():
    x = 333  # ローカル変数
    print(x)

use_global()  # 10
use_local()   # 333 / 関数内に同じ名前のローカル変数がある場合は、ローカル変数が優先される。
````

#### # グローバル変数を直接参照するより、引数として値を渡す方が再利用しやすく、処理の関係も分かりやすい。

````python
def print_value(value):
    print(value)

print_value(10)
````

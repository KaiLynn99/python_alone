## 1. ラムダ関数の構文

- ラムダ式は、関数名を定義せずに関数を作成するための式である。

- ラムダ式の結果として、関数オブジェクトが得られる。 

- したがって、ラムダによって生成された関数も第一級関数として扱われる。

````python
square = lambda x: x * x    # ラムダ式で作った関数を変数に代入する
print(square(5))
````

#### ※ Pythonには、次のような文と式がある。

条件文 / 条件式（三項演算子）

繰り返し文 / 内包表記

関数定義文 / ラムダ式

#### # ラムダ式からは関数オブジェクトが得られ、
#### その関数を呼び出すと、コロンの後ろの式の結果が返される。

```python
lambda 引数: 戻り値となる式
```

- `引数`：関数が受け取る値
- `戻り値となる式`：関数を呼び出したときに計算される式
- `ラムダ式全体`：関数オブジェクトを生成する式

````python
# 通常の関数の定義
def add1(a, b):
    return a + b

# ラムダ関数
add2 = lambda a, b : a + b    # add2には、"lambda a, b : a + b" というラムダ式が返した関数（オブジェクト）が代入される


r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)
````

````python
def make(n, func): 
    return func(n)

def f3(n):
    return "偶数" if n % 2 == 0 else "奇数" 

r = make(10, f3)
print(r)
````

## 2. ラムダ関数の即時呼び出し
- ラムダ式の結果値に () 呼び出し演算子を使用すると、ラムダ式が呼び出される。

````python
print((lambda x, y : x * y)(5, 10))
````

## 3. ラムダ関数は、他の関数の引数（パラメータ）としてよく使われる

````python
def test(fun):
    return fun(2) + 1000

def square(x):
    return x*x

def square2(x):
    return x-x

r1 = test(square)
print(r1)

r2 = test(square2)
print(r2)

r4 = test(lambda x: x*100)
print(r4)
````

#### # データサイエンスでよく使われる高階関数の種類

• コレクションと関数を引数として受け取り、<br>
コレクションから要素を一つずつ取り出して関数に渡す高階関数

````python
def ds_fun(data, fun):
    for d in data:
        fun(d)

numbers = [1, 2, 3]

ds_fun(numbers, lambda x: print(x * 10))
````

````python
s = [('カオナシ', 90), ('カルシファー', 85), ('バロン', 100)]

r = sorted(s, key=lambda x: x[1], reverse=True)

print(r)
````

• 通常の関数を使っても結果は同じである。

````python
s = [('カオナシ', 90), ('カルシファー', 85), ('バロン', 100)]

def score(x): 
    return x[1]

r = sorted(s, key=score, reverse=True)

print(r)
````

## 4. ラムダを引数に取る代表的な組み込み関数

• 関数（主にラムダ）を引数として受け取る関数を高階関数（Higher-order Function）と呼ぶ。

| 関数 | 関数・ラムダ式の役割 | 結果 |
|---|---|---|
| `map(func, data)` | 各要素を変換する | 変換された要素を持つ `map` オブジェクトを返す |
| `filter(func, data)` | 条件を判定する | 条件を満たす要素を持つ `filter` オブジェクトを返す |
| `sorted(data, key=func)` | 並べ替えの基準を指定する | 並べ替えられたリストを返す |
| `min(data, key=func)` | 最小要素を判断する基準を指定する | 基準値が最も小さい要素を返す |
| `max(data, key=func)` | 最大要素を判断する基準を指定する | 基準値が最も大きい要素を返す |
| `any(iterable)` | `True` の要素が一つでもあるか確認する | `True` または `False` を返す |
| `all(iterable)` | すべての要素が `True` であるか確認する | `True` または `False` を返す |

#### 1) map() - すべての要素を変換する

````python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, nums))

print(result)
````

#### 2) filter() - 条件に一致する要素のみを選択

````python
nums = [1, 2, 3, 4, 5]

result = list(filter(lambda x: x % 2 == 0, nums))

print(result)
````

#### 3) sorted() - ソート基準の指定

````python
names = ["banana", "apple", "kiwi"]

result = sorted(names, key=lambda x: len(x), reverse=True)

print(result)
````

##### * sorted() でラムダ式を使用すると、データサイエンスでよく使われる [dict, dict, ...] というデータ構造を、任意のキーで並べ替えることができる。

````python
students = [
    {"名前": "カブ", "年齢": 31, "点数": 90},
    {"名前": "かまじい", "年齢": 72, "点数": 85},
    {"名前": "ドーラ", "年齢": 64, "点数": 95}
]

# "名前"で昇順に並べ替え
result1 = sorted(students, key=lambda x: x["名前"])
print(result1)

# "年齢"の値を基準に昇順でソート
result2 = sorted(students, key=lambda x: x["年齢"])
print(result2)

# "得点"で降順に並べ替え
result3 = sorted(students, key=lambda x: x["点数"], reverse=True)
print(result3)
````

##### + 複数のキーに基づいて並べ替えることも可能だ。

````python
students = [
    {"名前": "ジジ", "クラス": 2, "点数": 90},
    {"名前": "マルクル", "クラス": 1, "点数": 90},
    {"名前": "リン", "クラス": 1, "点数": 95}
]

# 第1順位：クラス、第2順位：点数
result4 = sorted(students, key=lambda x: (x["クラス"], x["点数"]))
print(result4)
````

#### 4) min() - 基準に基づいて最小値を探す

````python
names = ["banana", "apple", "kiwi"]

result = min(names, key=lambda x: len(x))

print(result)
````

#### 5) max() - 基準に基づいて最大値を探す

````python
names = ["banana", "apple", "kiwi"]

result = max(names, key=lambda x: len(x))

print(result)
````

#### 6) any() - 条件を満たすものが1つでも存在するかどうかを確認する

````python
nums = [1, 3, 5, 8]

result = any(map(lambda x: x % 2 == 0, nums))

print(result)
````

#### 7) all() - すべてが条件を満たしているかを確認する

````python
nums = [2, 4, 6, 8]

result = all(map(lambda x: x % 2 == 0, nums))

print(result)
````

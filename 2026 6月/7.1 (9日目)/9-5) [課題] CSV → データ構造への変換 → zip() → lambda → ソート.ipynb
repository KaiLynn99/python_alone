## 問題 2

#### 生徒の

#### 名前リスト 

#### 国語リスト 

#### 英語リスト 

#### をそれぞれ保存した後、出力しなさい。

#### 出力例）

```python
print(names) # ['キキ', 'アシタカ', ...]
print(kors)
print(engs)         
```

````python
names = []
kors = []
engs = []

import csv 

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader)
   
    for line in reader: 
        names.append(line[0])
        kors.append(line[1])
        engs.append(line[2])

print(names)
print(kors)
print(engs)
````

## 第一級関数（First-Class Function）とは、関数をオブジェクトと同じように扱えるという性質を指す。

- 具体的には、関数を次のように扱うことができる。

1) 変数に代入できる。

2) 関数の引数（argument）として渡すことができる。

3) 関数の戻り値（return）として使用できる。

4) データ構造（list、dictなど）に格納できる。

#### 1. 変数に代入する

```python
def hello():
    print("Hello")

f = hello   # 関数そのものを変数に代入する（括弧なし）

f()
```

2. 関数の引数として渡す

```python
def square(x):
    return x * x

def test(fun):
    return fun(5)

print(test(square))
```

3. 関数を戻り値として返す

```python
def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()
```

4. データ構造に格納する

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```

## 1. enumerate() : 要素とインデックスを同時に取り出す

```python
names = ["キキ", "アシタカ", "トトロ"]

for i, name in enumerate(names):
    print(i, name) 
```

## 2. reversed() : 要素の順番を逆にする（元のリストは変更しない）

```python
nums = [1, 2, 3]

print(list(reversed(nums)))
print(nums)
```

## 3. any() : 一つでもTrueと判定される要素があればTrueを返す
#### (数値の場合、0はFalse、0以外はTrueと判定される)

```python
nums = [0, 0, 3]

print(any(nums))
```

## 4. all() : すべての要素がTrueと判定される場合にTrueを返す

```python
nums = [0, 2, 3]

print(all(nums))
```

## 5. zip() : 複数の反復可能オブジェクトから、同じ位置の要素をまとめて取り出す
## (要素数が異なる場合は、最も短いオブジェクトに合わせて処理される)

#### 1) zip -> list

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]
grades = ["A", "B", "A"]

result = zip(names, scores, grades)

print(list(result))
```

- ループ内で使用

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]

for name, score in zip(names, scores):
    print(name, score)
```

#### 2) zip -> dict

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]

result = dict(zip(names, scores))

print(result)
```

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]

result = {name : score for name, score in zip(names, scores)}

print(result)
```

- 3つの要素をそのままdict()に渡す場合<br>
(`dict()`は、各要素を「キーと値」の2要素の組として受け取る)<br>
(そのため、1つの要素に3つの値が含まれている場合は辞書に変換できない)

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]
gender = ["F", "M", "F"]

result = dict(zip(names, scores, gender))   # error
```

- 3つの要素を辞書に変換する方法

```python
names = ["キキ", "アシタカ", "トトロ"]
scores = [90, 80, 100]
gender = ["F", "M", "F"]

result = dict(zip(names, zip(scores, gender)))

print(result)
```

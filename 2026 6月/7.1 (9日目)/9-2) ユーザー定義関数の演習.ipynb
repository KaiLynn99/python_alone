## 1. 戻り値のない関数

```python
def hello():
    """
    'こんにちは'を出力し、None値を返す関数 
    """
    print("こんにちは")

result = hello()
# 関数の呼び出しは式である。式は値を返す。戻り値のない関数を呼び出すと、None値が返される。
print(result) # resultにはNoneが代入されている。
```

#### ※ 関数名のすぐ下に記述する """ ... """ という文字列は、`ドックスリング`（Docstring、Documentation String）と呼ばれる。

- ドックスリングは、関数、クラス、モジュールの説明（ドキュメント）を作成する際に使用する。

- 通常のコメント（#）とは異なり、プログラムによって読み取ることができる。

- help() 関数や IDE（VSCode、PyCharm など）で自動的に表示される。

```python
help(hello)
```

## 2. 戻り値（returnキーワード）を持つ関数

- `return`文は、関数の実行を直ちに終了し、指定した値を関数を呼び出した位置に返す。
- `return`の後に値を指定しない場合は、`None`を返す。

### 例

- `return`：関数を終了し、`None`を返す。
- `return 7`：関数を終了し、`7`を返す。
- `return x + y`：関数を終了し、`x + y`という式の計算結果を返す。

```python
def add(a, b):
    return a+b

print(add(3, 5))

result = add(3, 5)
print(result)
```

```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x, y, z # 3つの値を個別に返すのではなく、タプルとしてまとめて返す。

result2 = calc(10, 5)
print(result2)

r1, r2, r3 = calc(10, 5) # アンパッキング
print(r2)
print(r3)
```

## 3. パラメータ（引数値）を受け取る関数

```python
def greet(name):
    print(f"{name}さん、こんにちは")

greet("ナウシカ")
greet("ハウル")

# 関数内の引数も、関数のコードブロック内でのみ使用可能なローカル変数である。
print(name) # NameError: name 'name' is not defined
```

```python
def fun1(a, b):
    print("パラメータ a の引数: ", a)
    print("パラメータ b の引数: ", b)
```

#### # 関数呼び出し時に引数の位置によってパラメータの値を決定

```python
fun1(1, 2)

fun1(10, 20)
```

#### # 関数呼び出し時にパラメータ名を指定して引数値を入力

```python
fun1(a=100, b=200)

fun1(b=777, a=999) # パラメータ名を指定して引数を入力する場合（キーワード引数）、順序は関係ありません。
```

## 4. デフォルト値のパラメータ

#### 関数を呼び出す際に引数を省略すると、関数の定義時にパラメータとして指定されたデフォルト値が使用される。

- デフォルト値を持つ引数のルール

1) デフォルト値のない仮引数は、デフォルト値のある仮引数より先に記述する。

```python
def greet(name, age=20):
    print(name, age)

greet("サン")
greet("ポルコ", 30)
```

2) デフォルトパラメータの後に通常のパラメータを記述してはならない。

<u>デフォルトパラメータは常に最後に記述する。</u>

```python
# エラーの発生 : SyntaxError: parameter without a default follows parameter with a default
def greet(name="サン", age):
    print(name, age)
```

- 2つ以上のデフォルトパラメータ

```python
def calc(a, b=10, c=20):
    print(a, b, c)

calc(1)
calc(1, 2)
calc(1, 2, 3)
```

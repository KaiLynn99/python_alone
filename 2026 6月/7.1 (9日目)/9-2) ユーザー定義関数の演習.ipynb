## 11. ユーザー定義関数の演習

1）共通部分を求める

```python
def my_intersect(a, b):
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

print(my_intersect([1, 2, 3], [2, 3, 5]))   # リストの共通要素
print(my_intersect((1, 2, 3), (2, 3, 5)))   # タプルの共通要素
print(my_intersect({1, 2, 3}, {2, 3, 5}))   # セットの共通要素
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))   # 辞書の共通するキー
```

2) 共通要素の数を求める 

```python
def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count

print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

3) 差集合を求める 

```python
def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result

print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

4) 和集合を求める

```python
def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)
    
    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
print(my_union({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

5) 要素が含まれているか確認する

```python
def my_contains(data, value):

    for e in data:
        if e == value:
            return True
    
    return False

print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

6) 特定の値の個数を数える

```python
def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1
    
    return count

print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

7) 重複を削除する

```python
def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)
    return result

print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```

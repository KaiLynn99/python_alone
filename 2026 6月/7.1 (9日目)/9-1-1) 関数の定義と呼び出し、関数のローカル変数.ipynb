## 1. 関数

- 関数とは、特定の処理を実行するように名前を付けて定義されたコードの集まりである。

- 必要な時に呼び出して実行し、引数として値を受け取り、処理結果を返すことができる。

- 繰り返し行われる処理を関数化することで、コードの重複が減り、再利用や保守が容易になる。

## 2. 関数の定義と呼び出し

- 関数の定義：def を使用して関数を作成すること

- 関数の呼び出し：関数名の後に () を付けて関数を実行すること。関数の呼び出しは`式`である。

```python
def add(a, b):
    return a + b
```

- def -> 関数を定義するキーワード
- add -> 関数名
- (a, b) -> 引数（Parameter）
- : -> 関数のコードブロックの開始
- return -> 関数の実行を終了し、指定した値を関数の呼び出し元に返す。

- 関数の呼び出し（Call）：定義された関数を実行することを関数の呼び出しという。

- 関数の呼び出しは`式`である。関数を呼び出したコード行には、戻り値が残される。戻り値がない場合は、None値が残される。

```python
result = add(10, 20)
print(result)

# 10, 20 - 引数(Argument)
# add(10, 20) - 関数の呼び出し
# resultには、関数呼び出し式の戻り値が代入される
```

## 3. 関数のコードブロック内の変数の範囲（ローカルスコープ）

- Pythonでは、関数（def）とクラス（class）を除き、ほとんどのコードブロックは新しい変数のスコープを作らない。
- したがって、if、for、while、try、withブロック内で作成した変数は、ブロックの外側でも使用できる。

```python
def test():
    x = 100
    print(x)

test()

print(x)   # NameError
```

- その他の文は、コードブロック内に新しい変数のスコープを作らない。

```python
if True:
    x = 10

print(x)
```

```python
for i in range(3):
    value = i * 10

print(i)
print(value)
```

```python
try:
    result = 100 / 10
except:
    pass

print(result)
```

## まとめ

| `try-except` | ❌ |

## 4. 戻り値のない関数

```python
def hello():
    """
    'こんにちは'を出力し、None値を返す関数 
    """
    print("こんにちは")

result = hello()
# 関数の呼び出しは式である。式は値を返す。戻り値のない関数を呼び出すと、None値が返される。
print(result) # resultにはNoneが代入されている。
```

#### ※ 関数名のすぐ下に記述する """ ... """ という文字列は、`ドックスリング`（Docstring、Documentation String）と呼ばれる。

- ドックスリングは、関数、クラス、モジュールの説明（ドキュメント）を作成する際に使用する。

- 通常のコメント（#）とは異なり、プログラムによって読み取ることができる。

- help() 関数や IDE（VSCode、PyCharm など）で自動的に表示される。

```python
help(hello)
```

## 5. 戻り値（returnキーワード）を持つ関数

- `return`文は、関数の実行を直ちに終了し、指定した値を関数を呼び出した位置に返す。
- `return`の後に値を指定しない場合は、`None`を返す。

### 例

- `return`：関数を終了し、`None`を返す。
- `return 7`：関数を終了し、`7`を返す。
- `return x + y`：関数を終了し、`x + y`という式の計算結果を返す。

```python
def add(a, b):
    return a+b

print(add(3, 5))

result = add(3, 5)
print(result)
```

```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x, y, z # 3つの値を個別に返すのではなく、タプルとしてまとめて返す。

result2 = calc(10, 5)
print(result2)

r1, r2, r3 = calc(10, 5) # アンパッキング
print(r2)
print(r3)
```

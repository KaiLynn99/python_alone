## 1. ユーザー定義モジュール

### \- モジュールのimport
##### (モジュール内の変数、関数、クラスなどを使用するには、import キーワードを使用する)

### \- 3つのimport方法

- mathモジュール全体を読み込む。使用する際は、必ずmath.を付ける。
```python 
import math
```

- mathモジュールをmという別名（alias）で使用する。
```python 
import math as m
```

- mathモジュールからsqrt関数だけを読み込む。math.を付けずにそのまま使用できる。
```python 
from math import sqrt
```

````python
import my_math
my_math.add(2,3)
````

````python
import my_math as mm
mm.add(2,3)
````

````python
from my_math import add
add(3,6)
````

````python
from my_math import add as a
a(3,6)
````

#### 1）数学に関する定数や関数を提供する`math`モジュール

````python
import math

print(math.sqrt(16))   # 平方根
print(math.pi)         # 円周率
print(math.floor(3.7)) # 小数点以下を切り捨てる
print(math.ceil(3.2))  # 小数点以下を切り上げる
````

#### 2）乱数を生成する`random`モジュール

````python
import random

print(random.random())          # 0以上1未満の実数をランダムに生成する
print(random.randint(1, 10))    # 1以上10以下の整数をランダムに生成する
print(random.choice([1, 2, 3])) # リストから要素をランダムに1つ選択する
````

#### 3）日付と時刻を扱う`datetime`モジュール、処理を一時停止する`time`モジュール

````python
import datetime

now = datetime.datetime.now()  # 現在の日付と時刻を取得する
print(now)

print(now.year)   # 年を取得する
print(now.month)  # 月を取得する
print(now.day)    # 日を取得する
````

````python
import time

print("開始")
time.sleep(2)  # 処理を2秒間停止する
print("終了")
````

#### 4) ファイル・フォルダ関連のOSモジュール

````python
import os

print(os.getcwd()) # 現在の作業フォルダ
print(os.listdir()) # 現在のフォルダ内の一覧
````

#### 5）システム情報に関する`sys`モジュール

````python
import sys  # Pythonのシステムに関する機能を提供するモジュール

print(sys.version)  # Pythonのバージョン
print(sys.path)     # モジュールを検索するパス
````

#### 6）コレクションに関する`collections`モジュール

````python
from collections import Counter

# 数学に関する機能のうち、組み込み関数にない機能はmathモジュールで提供される。
# コレクションを扱う機能のうち、組み込み関数にない機能はcollectionsモジュールで提供される。

data = ["a", "a", "x", "a", "z", "z", 4]

print(Counter(data))  
````

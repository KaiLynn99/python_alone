# 1．リストの特徴と活用

リストは、末尾へのデータの追加（`append()`）と、末尾からのデータの削除（`pop()`）を高速に行うことができる。

そのため、スタック（Stack）を実装するのに適している。


# 2．スタック（Stack）

`LIFO (Last In, First Out)`とは、後から追加されたデータが最初に取り出される仕組みである。

```text
          ┌───┐
Top  →    │ C │
          ├───┤
          │ B │
          ├───┤
Bottom →  │ A │
          └───┘

- スタックの操作 <br>

push：データを追加する → リストの append() を使用する

pop：データを取り出す → リストの pop() を使用する

例：リストを使ったスタックの実装

#### 1) push

````python
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)
````

#### 2) pop

````python
num = stack.pop()

print(num)
print(stack)
````

#### 3) Top確認

````python
stack = [10, 20, 30]

print(stack[-1])
print(stack)
````

#### 4) スタックが空かどうかを確認する

````python
stack = []

if len(stack) == 0:
    print("スタックは空です。")
else:
    print("データがあります。")
````

#### 5) Webブラウザの「戻る」機能を実装する

````python
stack = []

stack.append("Yahoo")
stack.append("Google")
stack.append("ChatGPT")

print("現在の閲覧履歴：", stack)
stack.pop()  # 現在のページをスタックから取り出す
print("戻った後のページ：", stack[-1])
````

#### 6)「皿（コイン）を積み重ねる」処理を実装する

````python
stack = []

stack.append("皿1")
stack.append("皿2")
stack.append("皿3")

while len(stack) > 0:
    print(stack.pop())
````

````python
plates = []

plates.append({"color":"red", "price":100})
plates.append({"color":"blue", "price":150})
plates.append({"color":"blue", "price":150})
plates.append({"color":"black", "price":200})
plates.append({"color":"red", "price":100})
plates.append({"color":"black", "price":200})
plates.append({"color":"black", "price":200})

print(plates)

total_price = 0 

while len(plates) > 0:
    p = plates.pop()
    total_price += p["price"]

print(f"¥{total_price:,}")
print(plates) # 処理完了後、スタックが空かどうかを確認する
````

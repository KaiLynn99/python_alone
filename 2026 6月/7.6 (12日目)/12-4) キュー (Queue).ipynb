- スタックは、末尾にデータを追加し、末尾から取り出す（LIFO）。

- キューは、末尾にデータを追加し、先頭から取り出す（FIFO）。

# 1．キュー（Queue）の概念

`FIFO (First In, First Out)`とは、先に追加されたデータが先に取り出される仕組みである。

待ち行列をイメージすると分かりやすい。

# 2．キューとスタックの比較

| スタック（Stack） | キュー（Queue） |
|---|---|
| LIFO | FIFO |
| 最後のデータを削除する | 最初のデータを削除する |
| `append()` ＋ `pop()` | `append()` ＋ `pop(0)` |
| 皿を積み重ねる | 銀行の待ち行列 |

# 3．リストによる実装

## 1) データを追加する（Enqueue）

````python
queue = []

queue.append("A")
queue.append("B")
queue.append("C")

print(queue)
````

## 2) データを取り出す（Dequeue）

````python
queue = ["A", "B", "C"]

print(queue.pop(0))
print(queue)
````

## 3) 銀行の番号札の例

````python
queue = []

queue.append("ハウル")
queue.append("ソフィー")
queue.append("カルシファー")

print("待ち人数：", queue)
print("入場：", queue.pop(0))
print("現在の待ち人数：", queue)
````

## 4) プリンターの出力キューの例

````python
queue = []

queue.append("文書1")
queue.append("文書2")
queue.append("文書3")

while len(queue) > 0:
    print("印刷：", queue.pop(0))
````

## 5) 料理の注文例

````python
queue = []

queue.append("ハンバーガー")
queue.append("ピザ")
queue.append("ラーメン")

print("注文一覧：", queue)

print("調理開始：", queue.pop(0))
print("調理開始：", queue.pop(0))

print("残っている注文：", queue)
````

# 4. Stack vs Queue

````python
stack = []

stack.append("A")
stack.append("B")
stack.append("C")

while len(stack) > 0:
    print(stack.pop(0), end= " ")
````

````python
queue = []

queue.append("A")
queue.append("B")
queue.append("C")

while len(queue) > 0:
    print(queue.pop(0), end= " ")
````

````python
queue = ["A", "B", "C"]
pop_index = 0
while pop_index < len(queue):
    print(queue[pop_index]) # ["A", "B", "C"]
    print(queue[pop_index+1:])
    pop_index += 1
print(queue)
````

## ※ キューの要素数が多い場合、リストの使用は適していない

```python
queue = []

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)
queue.append(50)

queue.pop(0)

上のコードは、リストをキューのように使用している。

しかし、`pop(0)`を実行すると、先頭の要素を削除した後、残りの要素をすべて一つずつ前に移動させる必要がある。

| 処理 | リストの状態 |
|---|---|
| 削除前 | `[10, 20, 30, 40, 50]` |
| `pop(0)`を実行 | 先頭の `10` を削除 |
| 削除後 | `[20, 30, 40, 50]` |

`20`、`30`、`40`、`50`をすべて一つずつ前に移動させるため、データ数が多くなるほど、リストの `pop(0)` は非効率になる。

- `pop(0)` の時間計算量：**O(n)**
- 要素数が多いキューを実装する場合は、リストよりも `collections.deque` を使用する方が効率的である。

## `deque` の `popleft()` を使用する場合

```python
from collections import deque

queue = deque([10, 20, 30, 40, 50])

queue.popleft()
```

上のコードは、`deque` をキューとして使用している。

`popleft()`を実行すると、先頭の要素を削除できる。  
リストの`pop(0)`とは異なり、残りの要素を一つずつ前に移動させる必要はない。

| 処理 | キューの状態 |
|---|---|
| 削除前 | `deque([10, 20, 30, 40, 50])` |
| `popleft()`を実行 | 先頭の `10` を削除 |
| 削除後 | `deque([20, 30, 40, 50])` |

- リストの`pop(0)`：残りの要素を前に移動させるため **O(n)**
- `deque`の`popleft()`：先頭の要素を効率的に削除できるため **O(1)**

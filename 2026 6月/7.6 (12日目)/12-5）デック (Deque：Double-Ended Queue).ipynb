## デック（Deque）の概念

デック（Deque：Double-Ended Queue）とは、両端からデータの追加と削除ができるデータ構造である。

前述したように、Pythonでキューを実装する場合も、リストより `deque` を使用する方が効率的である。

デックは、キューの代わりとして使用したり、要素を回転させたりする場合に利用される。

- 正の値を指定すると、要素を右方向に回転する。
- 負の値を指定すると、要素を左方向に回転する。

# 1．デックの操作

## `deque` の主な機能

| メソッド・属性 | 説明 |
|---|---|
| `append(x)` | 右端に要素を追加する |
| `appendleft(x)` | 左端に要素を追加する |
| `pop()` | 右端の要素を取り出して削除する |
| `popleft()` | 左端の要素を取り出して削除する |
| `rotate(n)` | 要素を回転する（`+`：右方向、`-`：左方向） |
| `maxlen` | デックの最大要素数を指定する |

## 1) デックの作成

```python
from collections import deque

dq = deque()

print(dq)
```

## 2) 右端にデータを追加する：`append()`

```python
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

## 3) 左端にデータを追加する：`appendleft()`

```python
from collections import deque

dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10)

print(dq)
```

## 4) 右端のデータを削除する：`pop()`

```python
from collections import deque

dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)
```

## 5) 左端のデータを削除する：`popleft()`

```python
from collections import deque

dq = deque([10, 20, 30])
x = dq.popleft()

print(x)
print(dq)
```

## 6) 両端からデータの追加・削除ができる。

```python
from collections import deque

dq = deque()

dq.append(10)
dq.append(20)
dq.appendleft(5)
dq.appendleft(1)

print(dq)

print(dq.pop())
print(dq.popleft())

print(dq)
```

## 7) 要素を回転させる：`rotate()`

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 右方向に1つ移動（回転）

print(dq) # deque([5, 1, 2, 3, 4])

# 次の処理と同じ
dq1 = deque([1, 2, 3, 4, 5])
x = dq1.pop()
dq1.appendleft(x)
print(dq1)
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(-2) # 左方向に2つ回転

print(dq)
```

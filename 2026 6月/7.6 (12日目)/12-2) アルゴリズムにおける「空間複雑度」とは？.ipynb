# 1．空間計算量

プログラムの実行中に、追加で使用するメモリ量を**空間計算量（Space Complexity）**という。

入力サイズが大きくなるにつれて、使用するメモリ量がどの程度増加するかを表す。

通常、Big-O記法を用いて表現する。


## 例

- 変数を1個だけ使用するプログラム  
  → 入力サイズが増えても、使用するメモリ量はほとんど変化しない  
  → **O(1)**

- 入力サイズと同じ数の要素を持つリストを作成するプログラム  
  → 使用するメモリ量が入力サイズに比例して増加する  
  → **O(n)**


## ※ 時間計算量と空間計算量

- **時間計算量**：処理時間が入力サイズに応じて、どの程度増加するかを表す。

- **空間計算量**：メモリ使用量が入力サイズに応じて、どの程度増加するかを表す。

## 1）空間計算量：O(1)

```python
n = int(input("数値を入力してください："))
result = n * 2

print(result)

- 追加で使用する変数は、n と result の2個だけである。

- 入力値が10でも100万でも、使用する変数の個数は変わらない。

- そのため、空間計算量は O(1) となる

## 2）空間計算量：O(n)

```python
n = int(input("数値の個数を入力してください："))

numbers = []

for i in range(n):
    numbers.append(i)

print(numbers)

```python
# O(n)
n = 1000000
numbers = list(range(n))

- 入力サイズ n が大きくなるほど、リストに格納する要素数も増加する。

- 使用するメモリ量が入力サイズに比例して増加するため、空間計算量は O(n) となる。

## 3) 空間計算量

```python
n = int(input("サイズ : "))

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)

```python
# O(n²)
n = 1000
matrix = [[0] * n for i in range(n)]

- `n = 3` の場合、データは  
  `[[0, 0, 0], [0, 0, 0], [0, 0, 0]]` となり、要素数は `3 × 3 = 9` 個となる。

- 入力サイズ `n` が大きくなるほど、要素数は `n²` に比例して増加する。

- そのため、空間計算量は **O(n²)** となる。

| 空間計算量 | 意味 | 例 |
|---|---|---|
| O(1) | メモリ使用量が一定 | 数個の変数 |
| O(n) | 入力サイズに比例してメモリが増加 | 1次元リスト |
| O(n²) | 入力サイズの二乗に比例してメモリが増加 | 2次元リスト |

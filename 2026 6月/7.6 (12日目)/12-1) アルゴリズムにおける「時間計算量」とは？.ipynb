## ※ 時間計算量と空間計算量

**`時間計算量`（Time Complexity）**：
入力サイズの増加に伴って、アルゴリズムの実行時間がどの程度増加するかを表す。

**`空間計算量`（Space Complexity）**：
入力サイズの増加に伴って、アルゴリズムが使用するメモリ量がどの程度増加するかを表す。

## ※ アルゴリズムとの関係

良いアルゴリズムとは、実行時間が`短く`（Time）、メモリ使用量も`少ない`（Space）アルゴリズムである。

ただし、この二つは**`トレードオフ`（Trade-off）の関係**になる場合が多い。

- メモリを多く使用すると、実行時間を短縮できる場合がある。
- メモリの使用量を抑えると、実行時間が長くなる場合がある。

# 1．時間計算量

`同じ問題であっても、使用するアルゴリズムによって処理回数や実行時間が大きく異なることを確認`する。(それぞれのアルゴリズムの性能差を比較する)

時間計算量（Big-O）が、プログラムの性能を予測するための重要な基準であることを理解する。

時間計算量とは、データ量が増加したときに、処理回数がどの程度増加するかを表す基準である。

1. 線形探索 → 目的のデータを探す
2. 最大値探索 → 繰り返し処理を使って値を比較する
3. バブルソート → ソートアルゴリズムを理解する
4. 実行時間の比較 → 時間計算量（Big-O）の必要性を実感する

# 2．アルゴリズム

## 1）線形探索 ― 繰り返し処理

| **実習1．線形探索（10分）** |
|---|
| &nbsp; |
| **テーマ：目的の数字を探す** |
| &nbsp; |
| リストの先頭から末尾まで、順番にデータを探索する。 |
| データ数が増えるほど、探索回数も増加することを確認する。 |
| データ数が10倍になると、比較回数も最大で10倍になる。 |
| &nbsp; |
| **学習する概念** |
| ・線形探索 |
| ・時間計算量：**O(n)** |

```python
numbers = [15, 8, 23, 7, 31, 18]
target = 23

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print("見つかった位置：", i)
        break # 見つかったら、残りの要素は比較せずに繰り返しを終了する

print("比較回数 :", count)
```

- データを先頭から一つずつ確認する。

- データが n 個ある場合、最大で n 回確認する。

- 処理回数がデータ数に比例して増加するため、線形時間（Linear Time）と呼ばれる。

- このアルゴリズムの時間計算量は O(n) である。

## 2．最大値を探す ― 繰り返し処理

## 実習2．最大値を探す（10分）

### テーマ：最も大きい値を探す

繰り返し処理を使用して、最大値を自分で探す。

すべてのデータを一度だけ順番に確認するアルゴリズムを実装する。

最大値を探すためには、すべてのデータを確認する必要がある。
そのため、このアルゴリズムの時間計算量も **O(n)** となる。

### 学習する概念

* 比較結果を更新しながら最大値を求める処理
* 時間計算量：**O(n)**

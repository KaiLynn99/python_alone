## ※ 時間計算量と空間計算量

**`時間計算量`（Time Complexity）**：
入力サイズの増加に伴って、アルゴリズムの実行時間がどの程度増加するかを表す。

**`空間計算量`（Space Complexity）**：
入力サイズの増加に伴って、アルゴリズムが使用するメモリ量がどの程度増加するかを表す。

## ※ アルゴリズムとの関係

良いアルゴリズムとは、実行時間が`短く`（Time）、メモリ使用量も`少ない`（Space）アルゴリズムである。

ただし、この二つは**`トレードオフ`（Trade-off）の関係**になる場合が多い。

- メモリを多く使用すると、実行時間を短縮できる場合がある。
- メモリの使用量を抑えると、実行時間が長くなる場合がある。

# 1．時間計算量

`同じ問題であっても、使用するアルゴリズムによって処理回数や実行時間が大きく異なることを確認`する。(それぞれのアルゴリズムの性能差を比較する)

時間計算量（Big-O）が、プログラムの性能を予測するための重要な基準であることを理解する。

時間計算量とは、データ量が増加したときに、処理回数がどの程度増加するかを表す基準である。

1. 線形探索 → 目的のデータを探す
2. 最大値探索 → 繰り返し処理を使って値を比較する
3. バブルソート → ソートアルゴリズムを理解する
4. 実行時間の比較 → 時間計算量（Big-O）の必要性を実感する

# 2．アルゴリズム

## 1）線形探索 ― 繰り返し処理

| **実習1．線形探索（10分）** |
|---|
| &nbsp; |
| **テーマ：目的の数字を探す** |
| &nbsp; |
| リストの先頭から末尾まで、順番にデータを探索する。 |
| データ数が増えるほど、探索回数も増加することを確認する。 |
| データ数が10倍になると、比較回数も最大で10倍になる。 |
| &nbsp; |
| **学習する概念** |
| ・線形探索 |
| ・時間計算量：**O(n)** |

```python
numbers = [15, 8, 23, 7, 31, 18]
target = 23

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print("見つかった位置：", i)
        break # 見つかったら、残りの要素は比較せずに繰り返しを終了する

print("比較回数 :", count)
```

- データを先頭から一つずつ確認する。

- データが n 個ある場合、最大で n 回確認する。

- 処理回数がデータ数に比例して増加するため、線形時間（Linear Time）と呼ばれる。

- このアルゴリズムの時間計算量は O(n) である。

## 2．最大値を探す ― 繰り返し処理

| **実習2．最大値を探す（10分）** |
|---|
| &nbsp; |
| **テーマ：最も大きい値を探す** |
| &nbsp; |
| 繰り返し処理を使用して、最大値を求める。 |
| すべてのデータを1回だけ走査するアルゴリズムを実装する。 |
| 最大値を求めるには、すべてのデータを確認する必要がある。そのため、このアルゴリズムの時間計算量も **O(n)** となる。 |
| &nbsp; |
| **学習する概念** |
| ・比較しながら最大値を更新する処理 |
| ・時間計算量：**O(n)** |

```python
numbers = [15, 8, 23, 7, 31, 18]

max_num = numbers[0]
count = 0

for i in numbers:
    count += 1
    if i > max_num:
        max_num = i

print("最大値 :", max_num)
print("比較回数 :", count)
```

- すべてのデータを一つずつ確認する必要がある。

- 途中で処理を終了することはできず、最後の要素まで比較する必要がある。  
  線形探索のように、アルゴリズム内で `break` を使用して処理を終了することはない。

- そのため、時間計算量は **O(n)** となる。

## 3）バブルソート ― 二重ループを使用するアルゴリズム

| **実習3．バブルソート（20分）** |
|---|
| &nbsp; |
| **テーマ：数値を昇順に並べ替える** |
| &nbsp; |
| 隣接する二つの値を比較し、必要に応じて入れ替える処理を実装する。 |
| データ数が増えるほど、比較回数が急激に増加することを確認する。 |
| &nbsp; |
| **学習する概念** |
| ・二重ループ |
| ・バブルソート |
| ・時間計算量：**O(n²)** |

```python
numbers = [5, 3, 8, 1, 4]   

# データ数：5個

# 1回目：4回比較

# 2回目：3回比較

# 3回目：2回比較

# 4回目：1回比較

count = 0

for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        count += 1
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i] 
            # temp = numbers[i]
            # numbers[i] = numbers[i+1]
            # numbers[i+1] = temp

print(numbers)
print("比較回数 :", count)
```

- バブルソートでは、繰り返し処理の中に別の繰り返し処理を使用する。

- 比較回数は、`(n-1) + (n-2) + ... + 1` となり、約 `n²` に比例して増加する。

- そのため、時間計算量は **O(n²)** となる。

- Big-Oは正確な処理回数ではなく、データ数の増加に伴って処理量がどのように増えるかを表す。

## 4）実行時間の比較

同じデータ数における O(n) と O(n²) の実行時間を比較し、時間計算量による性能差を確認する。

`データ数が増加するほど、アルゴリズムの性能差がどの程度大きくなるかを確認`する。

| **実習4．処理回数の比較（20分）** |
|---|
| &nbsp; |
| **テーマ：アルゴリズムの性能比較** |
| &nbsp; |
| 線形探索、最大値探索、バブルソートの処理回数を実際に出力して比較する。 |
| `time`モジュールを使用して実行時間を測定し、入力サイズによる性能差を確認する。 |
| &nbsp; |
| **［1］実行時間の測定方法** |
| ・`time.time()`を使用して、プログラムの実行時間を測定する方法を学ぶ。 |
| ・アルゴリズムの性能を数値で確認する方法を身につける。 |
| &nbsp; |
| **［2］線形探索（O(n)）** |
| ・データを先頭から末尾まで一つずつ確認するアルゴリズムを理解する。 |
| ・データ数が増加すると、実行時間もほぼ比例して増加することを確認する。 |
| &nbsp; |
| **［3］バブルソート（O(n²)）** |
| ・二重ループを使用するソートアルゴリズムを理解する。 |
| ・データ数が増加するほど、実行時間が急激に増加することを確認する。 |
| &nbsp; |
| **［4］時間計算量の比較** |
| ・同じデータ数でも、アルゴリズムによって実行時間が大きく異なる場合があることを確認する。 |
| ・時間計算量（Big-O）が、プログラムの性能を予測するための重要な基準であることを理解する。 |

```python
import random
import time

# 共通で使用するランダムデータを生成
original_numbers = random.sample(range(10000), 1000)

# 線形探索ではリストの内容を変更しないため、そのまま参照する
search_numbers = original_numbers

# バブルソートではリストの並び順を変更するため、コピーを作成する
sort_numbers = original_numbers.copy()
```

## 1) 線形探索（O(n)）

```python
# 実行時間の測定を開始
start = time.time()

# 最後の要素を探索対象に設定する（最悪の場合）
target = search_numbers[-1]

# 先頭から末尾まで一つずつ比較する
for n in search_numbers:
    if n == target:
        break

# 実行時間の測定を終了
end = time.time()

print("線形探索：", end - start)
```

## 2) バブルソート（O(n²)）

```python
# 実行時間の測定を開始
start = time.time()

# バブルソートを実行する
for j in range(len(sort_numbers) - 1):
    for i in range(len(sort_numbers) - 1 - j):
        if sort_numbers[i] > sort_numbers[i + 1]:
            sort_numbers[i], sort_numbers[i + 1] = (
                sort_numbers[i + 1],
                sort_numbers[i]
            )

# 実行時間の測定を終了
end = time.time()

print("バブルソート：", end - start)
```

- 1,000個のデータを処理した結果、線形探索は約0.00029秒、バブルソートは約0.122秒かかった。

- この実習で重要なのは、データ数が増えるほど **O(n)** のアルゴリズムが **O(n²)** のアルゴリズムよりも大幅に有利になるという点である。

- 線形探索はデータを1回確認するため **O(n)**、バブルソートは繰り返し比較するため **O(n²)** となる。データ数が増えるほど、両者の性能差は大きくなる。

## 1. whileループ

- while文は、条件式が真である間、処理を繰り返す。
- 条件式が偽になるとループを終了し、次の処理に進む。

```python
greet = "Hello"
while greet: # 文字列の場合、空白であればFalse、そうでなければTrue
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10 
# 複数割り当て（多重代入）
# 内部的には、右側のタプル (0, 10) をアンパックして、それぞれの変数に代入する

while a < b:
    print(a)
    a = a + 1 #a += 1
```

## 2. for

- リスト、タプル、文字列など、反復可能なオブジェクトの要素を1つずつ順番に処理する。

- 処理対象のオブジェクトから値を1つずつ取り出して変数に代入し、以下のコードブロックを繰り返し実行する。

```python
for a in [5, 10, 15]:
    print(a)
```

```python
for (a, b) in [(1, 2), (10, 20)]:
    print(a)
    print(b)
```

```python
for a, *b, c in  [(1, 2, 3, 4), (5, 6, 7, 8)]:
    print(a, b, c)
```

## 3. break, continue, pass, loop else

#### ループ制御文

| 文 | 役割 |
|---|---|
| `break` | ループを直ちに終了する。 |
| `continue` | 現在の処理を中断し、次の繰り返しに進む。 |
| `pass` | 何も処理せず、そのまま次に進む。 |
| `else` | `break`を使わずにループが正常終了した場合に実行する。 |

##### 1) break

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

print("ループ終了")
```

##### 2) continue

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i) 
```

##### 3) pass
##### (後で実装する処理をあらかじめ用意する場合、空のブロックにpassを使用する)

```python
def my_func():
    pass
```

```python
if i % 2 == 0 :
    pass
```

##### 4) else 
##### (主に「最後まで探しても見つからなかった場合」に使うが、`if`の`else`と混同されやすいため、使用頻度は低い)

```python
# breakがない場合

for i in range(3):
    print(i)
else:
    print("正常終了")
```

```python
# break がある場合

for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("正常終了") # breakが実行されたため、elseは実行されない。
```

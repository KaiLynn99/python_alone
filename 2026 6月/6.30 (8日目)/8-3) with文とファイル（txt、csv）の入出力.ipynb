#### TXTファイルとCSVファイルの違い

| 区分    | TXTファイル                     | CSVファイル               |
| ----- | --------------------------- | --------------------- |
| 目的    | 一般的なテキストの保存                 | 表形式のデータの保存            |
| 形式    | 自由な形式で記述可能                  | 行（Row）と列（Column）で構成   |
| 区切り文字 | なし（自由形式）                    | 通常はカンマ（`,`）           |
| 解析    | `split()`などを使用して直接処理する必要がある | `csv.reader()`が自動的に処理 |
| 使用例   | メモ、ログ、設定ファイル                | Excelデータ、データ分析        |

#### ※ with文

- with文は、ファイルやリソース（Resource：プログラムがシステムから借りて使用するもの）などを使用した後、自動的に後処理（クローズ）を行うための構文である。

```python
with リソースの解放が必要なオブジェクト as 変数:
    実行する処理
```

#### # with文を使用しない場合

1\) 通常通り手動で閉める場合

````python
f = open("students.csv", "r", encoding="utf-8")

print(f.closed)  # False：ファイルが開いている


f.close()
print(f.closed)  # True：ファイルが閉じている
````

2\) エラーにより close() が実行されない場合

````python
f = open("students.csv", "r", encoding="utf-8")
print(f.read())

print(10 / 0)   # error 発生
f.close()   # 手動で閉じる必要があり、エラーにより実行されない。 

# with文を使用しない場合は、`open()`でファイルを開いた後、`f.close()`を記述して手動で閉じる必要がある。

# ただし、`f.close()`が実行される前に例外が発生すると、ファイルが閉じられない可能性がある。
````

````python
print(f.closed)  # False：ファイルが閉じられていない

f.close()        # 確認後、手動で閉じる
print(f.closed)  # True : ファイルが閉じられている
````

#### # try-finally文によるリソースの解放

````python
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10 / 0)   # error 発生
finally:
    f.close()
    print(f.closed)
    print("リソースの解放に成功しました")

# 例外が発生してもfinallyブロック内の`f.close()`は必ず実行されるため、ファイルは正常に閉じられる。

# ただし、例外自体は処理されていないため、エラーメッセージは表示される。
````

#### # with文を使用する場合

````python
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())

print(f.closed)

# with文を使用すると、withブロックの終了時にファイルが自動的に閉じられる。

# そのため、`f.close()`を記述する必要はなく、withブロックの外で`f.closed`を確認すると`True`が出力される。
````

## 3. CSVファイルとして保存する
- csv.DictReader()で読み込んだdataを修正した場合は、csv.DictWriter()または通常のファイル書き込み処理を使用して、再度保存できる。

#### 1) 修正前

````python
import csv

with open("students2.csv", "r", encoding="utf-8") as f:
    print(f.read(), end="")
````

````python
import csv

# CSVファイルを読み込む
with open("students2.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

# 点数を修正する
    for row in data:
        row["点数"] = int(row["点数"]) + 5

# 修正したデータを新しいCSVファイルに保存する
with open("students2_result.csv", "w", newline="", encoding="utf-8") as f:

    # data[0].keys() → 最初の辞書のキーをCSVの列名として使用する
    writer = csv.DictWriter(f, fieldnames=data[0].keys())

    # CSVの1行目に列名を書き込む
    writer.writeheader()

    # data内のすべての辞書を複数行として書き込む
    writer.writerows(data)

# 保存したCSVファイルの内容を確認する
with open("students2_result.csv", "r", encoding="utf-8") as f:
    print(f.read(), end="")
````

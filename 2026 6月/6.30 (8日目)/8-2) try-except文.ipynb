## 1. try-except文

- 例外（Exception）が発生しても、プログラムが終了することなく、所望の処理を実行することができる。

- 例外はコードのエラーではないが、プログラムの実行中にユーザーの入力値や実行環境などによって発生するエラーである。

```python
try:
    # 例外が発生する可能性のある処理
    ...

except 例外の種類:
    # 例外処理
    ...

else:
    # 例外が発生しなかった場合に実行（任意）
    ...

finally:
    # 例外の有無にかかわらず必ず実行（任意）
    ...
```

#### # try-except文を使用すると、tryブロック内で例外が発生した場合にexceptブロックが実行され、プログラムは終了せず、その後の処理を継続する。

````python
try:
    num = int(input("数字を入力してください：")) ** 2
    print(num)
except:
    print("数字ではありません。")

print("次の処理を実行します...")
````

#### # except文では、指定した特定の例外のみを処理する。

````python
try:
    num = int(input("数字を入力してください："))
    result = 10 / num
    print(result)

except ValueError:
    print("数字を入力してください。")

except ZeroDivisionError:
    print("0で割ることはできません。")
````

## 2. finallyの使用

- 例外の発生有無にかかわらず、必ず実行される。

````python
try:
    print(10 / 2)
except:
    print("エラーが発生しました")
finally:
    print("プログラムを終了します")
````

## 3. else の使用

​- 例外が発生しない場合に実行

```text
try
│
├── 例外発生 ─────▶ except
│
└── 例外なし ─────▶ else
          │
          ▼
      finally（必ず実行）
```

````python
try:
    num = int(input("数字を入力してください : "))
    result = 100 / num

except ValueError:
    print("数字を入力する必要があります。")

except ZeroDivisionError:
    print("0で割ることはできません。")

else:
    print("計算結果 :", result)

finally:
    print("プログラムを終了します。")
````

````python
try:
    with open("test.txt", "r", encoding="utf-8") as f:
        text = f.read()

except FileNotFoundError:
    print("ファイルがありません。")

else:
    print(text)

finally:
    print("ファイルの読み取り処理が完了しました。")
````

```python
# 1. 不変性


# Pythonの組み込みオブジェクトのうち、値だけを持つオブジェクト（数値、bool）はその値を変更できない。
# Pythonの文字列も変更できない。

s = "python"
print(s[0])
s[0] = "C" # error
```

```python
# Pythonの型の中でlist、dict、setは可変型である。(オブジェクト生成後に内部データを変更できるオブジェクト)

nums = [1, 2, 3]
nums[0] = 100
print(nums)
```

```python
# 文字列は変更できないため、文字列の演算は新しい文字列を生成して返す。

s = "python"

new_s = "C" + s[1:]

print(s)       # python
print(new_s)   # Cython
```

```python
# 2. 文字列変換組み込み関数（Python言語に含まれている関数）


# 整数（int）と文字列（str）は直接結合できないため、str()で文字列に変換する必要がある。

year = 2026
month = 6
day = 22

date = str(year) + "-" + str(month) + "-" + str(day)

print(date)
```

```python
a = "123"
b = "10"

print(a + b) # 12310

result = int(a) + int(b)

print(result)
```

```python
# 3. 文字列メソッド


# メソッドは特定のオブジェクトに属し、そのオブジェクトにのみ動作する関数であり、呼び出し式で実行される。
# (* 文字列は不変オブジェクトである。 文字列メソッドは元の文字列を変更するものではない。)

s = "python".upper()
print(s)
s2 = s.lower()
print(s2)
```

```python
# dir 関数

# dir関数は、引数として渡されたオブジェクトで利用可能なすべての属性を返す。

s = "python"
print(dir(s))
```

```python
# - 文字列の分割

data = "apple,banana,orange"

result = data.split(",")

print(result) # ['apple', 'banana', 'orange']
```

```python
# - 文字列の結合

fruits = ["apple", "banana", "orange"]

result = ", ".join(fruits)

print(result)
```

```python
# - 電話番号の接続

phone1 = "010"
phone2 = "1234"
phone3 = "5678"

phone = "-".join([phone1, phone2, phone3])

print(phone)
```

```python
# 1) - メールからIDを抽出

email = "python@test.com"

user_id = email.split("@")[0]  # @を基準に分離した後、最初の要素（インデックス0）を抽出

print(user_id)
```

```python
# 2) - ファイル名から拡張子を抽出

filename = "report.xlsx"

ext = filename.split(".")[-1]  # . を基準に分割し、最後の要素（インデックス -1）を抽出

print(ext)
```

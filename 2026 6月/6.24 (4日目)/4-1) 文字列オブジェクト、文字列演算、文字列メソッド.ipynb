```python
# 1) インデックス演算

s = "python"

print(s[0])
print(s[1])
print(s[-1])
```

```python
# 2) 文字列のループ

for ch in "python":
    print(ch)
```

```python
# 3) スライシング演算

# シーケンス[start:stop:step] -> 順序/長さのあるデータの集合

s = "python"

print(s[0:3])
print(s[2:5])
print(s[::-1])
```

```python
# 4) 接続演算

a = "py"
b = "thon"

print(a + b)
```

```python
# 5) 繰り返し演算

print("ha" * 3)
```

```python
# 6) メンバーシップの計算

print("py" in "python")
print("java" not in "python")
```

```python
# 7) 長さの計算

print(len("python"))
```

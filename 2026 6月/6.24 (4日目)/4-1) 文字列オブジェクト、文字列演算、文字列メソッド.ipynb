## 1. Pythonの文字列（str）はシーケンス（Sequence）型

- シーケンスとは、順序や長さを持つデータの集合である。

- インデックス操作が可能である。

- スライシングが可能である。

- 反復処理（for）が可能である。

- 長さを確認できる（len）集合である。

#### 1) インデックス演算 

```python
s = "python"

print(s[0])
print(s[1])
print(s[-1])
```

#### 2) 文字列のループ

```python
for ch in "python":
    print(ch)
```

#### 3) スライシング演算

```python
# シーケンス[start:stop:step]
s = "python"

print(s[0:3])
print(s[2:5])
print(s[::-1])
```

#### 4) 連結演算

```python
a = "py"
b = "thon"

print(a + b)
```

#### 5) 反復演算

```python
print("ha" * 3)
```

#### 6) メンバーシップ演算

```python
print("py" in "python")
print("java" not in "python")
```

#### 7) 長さ演算

```python
print(len("python"))
```

## 2. 不変性


- Pythonの組み込みオブジェクトのうち、単一の値だけを持つオブジェクト（数値、bool）はその値を変更できない。

#### ※ Pythonの文字列も変更できない。

```python
s = "python"
print(s[0])
s[0] = "C" # error
```

#### ※ Pythonの型の中でlist、dict、setは可変型である。(オブジェクト生成後に内部データを変更できるオブジェクト)

```python
nums = [1, 2, 3]
nums[0] = 100
print(nums)
```

#### ※ 文字列は変更できないため、文字列の演算は新しい文字列を生成して返す。

```python
s = "python"

new_s = "C" + s[1:]

print(s)       # python
print(new_s)   # Cython
```

## 4. 文字列変換組み込み関数（Python言語に含まれている関数）

```python
year = 2026
month = 6
day = 22
date = year + "-" + month + "-" + day # TypeError: unsupported operand type(s)
```

#### # 整数（int）と文字列（str）は直接結合できないため、str()で文字列に変換する必要がある。

```python
year = 2026
month = 6
day = 22

date = str(year) + "-" + str(month) + "-" + str(day)

print(date)
```

```python
a = "123"
b = "10"

print(a + b) # 12310

result = int(a) + int(b)

print(result)
```

## 5. 文字列メソッド

- メソッドは特定のオブジェクトに属し、そのオブジェクトにのみ動作する関数であり、呼び出し式で実行される。

- メソッド呼び出し式
オブジェクト.メソッド名(引数) # 引数、パラメータ、入力値、…などの類義語 

```python
s = "python".upper()
print(s)
s2 = s.lower()
print(s2)
```

### ※ dir 関数

- dir関数は、引数として渡されたオブジェクトで利用可能なすべての属性を返す。

```python
s = "python"
print(dir(s))
```

### ※ よく使われる文字列メソッド

#### - 大文字・小文字の変換

```python
text = "Python Programming"

print(text.upper())      # すべて大文字

print(text.lower())      # すべて小文字

print(text.capitalize()) # 先頭文字のみ大文字

print(text.title())      # 各単語の先頭文字を大文字
```

#### - 空白の削除

```python
text = "   Python   "

print(text.strip())   # 両端の空白を削除

print(text.lstrip())  # 左側の空白を削除

print(text.rstrip())  # 右側の空白を削除
```

#### - 文字列検索

```python
text = "Python Programming"

print(text.find("Pro"))    # 指定した文字列が見つかったら、最初の位置を返す
print(text.find("Java"))    # 指定した文字列が見つからなかったら、-1を返す
print(text.startswith("Python"))    # 指定した文字列で始まっていたら、Trueを返す
print(text.endswith("ing"))    # 指定した文字列で終わっていたら、Trueを返す
```

#### - 文字列の置換

```python
text = "I like Java"

print(text.replace("Java", "Python"))    # 指定した文字列が見つかったら、別の文字列に置き換える
```

#### - 文字列の分割

```python
data = "apple,banana,orange"

result = data.split(",")    # 指定した区切り文字で文字列を分割し、リストにする

print(result) # ['apple', 'banana', 'orange']
```

#### - 文字列の結合

```python
fruits = ["apple", "banana", "orange"]

result = ", ".join(fruits)    # 指定した区切り文字で要素を結合し、1つの文字列にする

print(result)
```

#### - 文字列の数を数える

```python
text = "banana"

print(text.count("a"))    # 指定した文字列が含まれる回数を返す
print(text.count("na")) 
```

#### - 電話番号の接続

```python
phone1 = "010"
phone2 = "1234"
phone3 = "5678"

phone = "-".join([phone1, phone2, phone3])    # 指定した区切り文字で複数の文字列を結合する

print(phone)
```

#### - メールからIDを抽出

```python
email = "python@test.com"

user_id = email.split("@")[0]  # @を基準に分離した後、最初の要素（インデックス0）を抽出

print(user_id)
```

#### - ファイル名から拡張子を抽出

```python
filename = "report.xlsx"

ext = filename.split(".")[-1]  # . を基準に分割し、最後の要素（インデックス -1）を抽出

print(ext)
```

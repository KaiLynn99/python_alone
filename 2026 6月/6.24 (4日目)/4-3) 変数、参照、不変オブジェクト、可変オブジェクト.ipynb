## 1. 変数


#### • 変数はオブジェクトが最初に割り当てられるときに生成される 
#### • (追加割り当ては、すでに生成された変数の参照を変更する)

#### • 変数がオブジェクトを指すリンクを「参照」と呼ぶ。

#### • 変数は任意の型のオブジェクトを参照でき、使用される前に代入されなければならない。

#### • オブジェクトは、自分が意味する「値」と「機能」を持つ割り当てられたメモリの断片である。

`````python
a = 3
a = "hello"
`````

#### 1) 数字3を表す新しいオブジェクトの生成

#### 2) 変数aがまだ存在しない場合の生成

#### 3) 変数aを新しいオブジェクト3に接続（代入、代入、参照）

#### 4) 文字列 "hello"を表す新しいオブジェクトの生成

#### 5) 変数 a をオブジェクト "hello" に接続

## 2. 共有参照


#### • 不変オブジェクトと可変オブジェクトは、共有参照が発生したときに動作が異なる。

#### 1) 不変オブジェクトの共有参照


##### • int、float、bool、str、tuple、frozenset などが不変オブジェクトである。

#### # 数字

`````python
a = 10
b = a

print(a, b) # 10 10
print(a is b) # Trueのようなオブジェクト参照
`````

#### • 同じオブジェクトを参照 - 共有参照

```text
a ──┐
    ├──> 10
b ──┘
```

#### • 値の変更

`````python
a = 20

print(a)
print(b)
`````

#### • 数値は不変オブジェクトであるため、値を変更するのではなく、新しい20を作成し、aは20を参照する。
#### - 不変オブジェクトの共有参照の値を変更する際は、新しいオブジェクトを作成してから参照する。

```text
a ───> 20
b ───> 10
```

#### # 文字列も不変オブジェクトである

`````python
s1 = "Python"
s2 = s1

print(s1 is s2) # True
print(s1 == s2) # True
`````

```text
s1 ──┐
     ├──> "Python"
s2 ──┘
```

`````python
s1 = s1 + "3"

print(s1)
print(s2)
`````

```text
s1 ───> "Python3"
s2 ───> "Python"
```

#### 2) 可変オブジェクトの共有参照


##### # list、dict、set、ユーザー定義クラスは可変オブジェクトである

`````python
nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) # True
`````

```text
nums1 ──┐
        ├──> [10, 20, 30]
nums2 ──┘    [0]  [1]  [2]
```

`````python
nums1.append(40)

print(nums1) # [10, 20, 30, 40]
print(nums2) # [10, 20, 30, 40]
`````

```text
nums1 ──┐
        ├──> [10, 20, 30, 40]
nums2 ──┘
```

#### 3) 等価（同じ値）比較（==）と同一オブジェクト（同じ参照）比較（is）


##### # 変数を再起動すると、オブジェクトが修正されるわけではなく、変数が異なるオブジェクトを参照することになる。
##### (値修正ができない不変客体をメモリ中に多数作るのはメモリ浪費）

`````python
a = 100
b = 100

print(a == b) # True 値の比較
print(a is b) # Trueオブジェクト比較
print()
`````

`````python
s1 = "Python"
s2 = "Python"

print(s1 == s2) # True
print(s1 is s2) # True
`````

```text
s1 ──┐
     ├──> "Python"
s2 ──┘
```

##### # 次の場合、異なるオブジェクトを参照しているが、オブジェクトの値が同じ場合である。

`````python
s1 = "Python"
s2 = "Python3"
s1 = s1 +"3"

print(s1 == s2) # True
print(s1 is s2) # False
`````

```text
s1 ───> "Python3"
s2 ───> "Python3"
```

#### # リスト
##### # 可変オブジェクトの場合、値が変更される可能性がある、
##### • たとえ、同じ値のオブジェクトを代入して変数を作成しても、それぞれ異なるオブジェクトを参照することになる。

`````python
a = [1, 2, 3]
b = a

print(a == b) # True
print(a is b) # True

# 組み込み関数 id() を使えば、同じオブジェクトを参照しているかどうかを確認できる

print(id(a))
print(id(b))
`````

````text
a ──┐
    ├──>  [1, 2, 3]
b ──┘
````

`````python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b) # True
print(a is b) # False

# 組み込み関数 id() を使えば、同じオブジェクトを参照しているかどうかを確認できる

print(id(a))
print(id(b))
`````

```text
a ───> [1, 2, 3]
b ───> [1, 2, 3]
```

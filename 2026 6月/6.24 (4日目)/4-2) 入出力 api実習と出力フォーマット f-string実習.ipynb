```python
# f-string

# オブジェクトを文字列フォーマットで出力するフォーマット文法。 すべてのオブジェクトを文字列に変換し、フォーマットする。
```

```python
# 1) 文字入力、整列

name = "Python"

print(f"{name:*<15}")
print(f"{name:*^15}")
print(f"{name:*>15}")

w = 15
print(f"{name:*>{w}}")
```

```python
# 2) 数字を埋める

num = 123

print(f"{num:08}")
print(f"{num:010}")
```

```python
# 3) 符号を含めて出力

a = 150
b = -50

print(f"{a:+}")
print(f"{b:+}")
```

```python
# 4) 正の数は空白、負の数のみ符号を含めて出力

c = 100
d = -100

print(f"{c: }")
print(f"{d: }")
```

```python
# 5) 千単位の区切り記号を出力

sales = 1234567890

print(f"{sales:,}")
```

```python
# 6) 千単位の区切り記号 + 桁数の調整

sales2 = 1234567

print(f"{sales2:>15,}")
print(f"{sales2:<15,}")
```

```python
# 7) 小数点の四捨五入

pi = 3.1415926535

print(f"{pi:.2f}")
print(f"{pi:.4f}")
```

```python
# 8) 指数表記法

num = 123456789

print(f"{num:e}")
print(f"{num:E}")
```

```python
# 9) パーセンテージ表示

rate = 0.8746

print(f"{rate:.0%}")
print(f"{rate:.1%}")
print(f"{rate:.2%}")
```

```python
# 10) 2進数、8進数、16進数

num = 255

print(f"{num:b}")
print(f"{num:o}")
print(f"{num:x}")
print(f"{num:X}")
```

```python
# 11) 日付フォーマット

# Python言語に含まれるオブジェクト/関数であっても、組み込みオブジェクト/関数とモジュール内に含まれるオブジェクト/関数に区別される。
# モジュールに含まれるものはimportが必要。

from datetime import datetime
today = datetime.now()

print(f"{today:%Y-%m-%d}")
print(f"{today:%H:%M:%S}")
print(f"{today:%Y/%m/%d %H:%M}")
```

```python
# 12) 変数と値を同時に出力

name = "Park"
age = 27

print(f"{name = }")
print(f"{age = }")
```

```python
# 13) 計算結果

a = 10
b = 20

print(f"{a} + {b} = {a+b}")
print(f"{a*b = }")
```

```python
# 14) 表形式で出力

name = "パク キョンピル"
age = 27
score = 99.9

print(f"{'이름':<10}{'나이':<10}{'점수':<10}")
print(f"{name:<10}{age:<10}{score:<10.1f}")
```

```python
# 15) 複合

money = 3500000.75

print(f"{money:>15,.2f}")
```

```python
# 16) 報告書の形態

income = 123456789
expense = 98765432
profit = income - expense

print(f"{'売り出し':<10}: {income:>15,}")
print(f"{'費用':<10}: {expense:>15,}")
print(f"{'利益':<10}: {profit:>15,}")
```

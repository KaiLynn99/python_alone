- 正規表現はデータ分析において重要であるが、複雑な文字列パターンを自分で一から作成する機会は多くない。

- まずは、簡単でよく使われるパターンに慣れておけばよい。複雑なパターンが必要になった場合は、その都度検索して使用する。

- 正規表現を自分で作成できなくても、検索したパターンを見て、`「これはどのような文字列を検索するパターンなのか」`を読み取ることができれば十分である。

## 1. `re` モジュールでよく使用する関数

`re` モジュールの関数には、文字列メソッドと機能が重複するものもある。

- 具体的な文字列を検索する場合  
  → 文字列メソッドを使用する方が簡単な場合がある。

- 特定のパターンを持つ文字列を検索する場合  
  → `re` モジュールの関数を使用する方が適している。

| `関数` | `主な機能` |
|---|---|
| `match()` | `文字列の先頭`から、`パターンに一致するか`を確認する |
| `search()` | `文字列全体`から、`最初に一致するパターン`を検索する |
| `findall()` | `一致したすべての結果`を、`リスト`で返す |
| `finditer()` | `一致したすべてのMatchオブジェクト`を、`イテレータ`で返す |
| `sub()` | `パターンに一致した文字列`を、`指定した文字列`に`置換する` |
| `split()` | `パターンに一致した部分`を`区切り`として、`文字列を分割`する |

## 1) `re.match()`：文字列の先頭から、パターンに一致するかを確認する。

```python
import re

text = "Python"
result = re.match(r"Py", text)

print(result)
```

```python
import re

text = "Python"
result = re.match(r"th", text)

print(result) # match() は文字列の先頭からのみ、パターンに一致するかを確認する
```

## 2) `re.search()`：文字列全体から、パターンに一致する部分を検索する。

```python
import re

text = "Python"

print(re.search(r"th", text))
```

## 3) `re.findall()`：パターンに一致した`すべての結果`を、`リスト`で返す。

```python
import re

text = "apple banana orange"
result = re.findall(r"a", text)

print(result)
```

```python
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 数字1文字、\d+ 連続する数字が1文字以上

print(numbers)
```

## 4) `re.sub()`：パターンに一致した部分を、指定した文字列に置換する。

```python
import re

text = "A10 B20 C30"
result = re.sub(r"\d", "", text) # # 数字文字の削除

print(result)
```

# 1．正規表現のパターン
 
> 正規表現のパターンとは、**文字列の規則を表す式**である。


## 例：電話番号のパターン

### 1）対象となる文字列

次のような電話番号を探すとする。

```text
010-1234-5678
```


### 2）文字列の規則

この電話番号は、次の規則で構成されている。

```text
「010-」＋ 数字4桁 ＋「-」＋ 数字4桁
```


### 3）正規表現で表す

この規則を正規表現で表すと、次のようになる。

```text
010-\d{4}-\d{4}
```

# 2．よく使われる正規表現のメタ文字
  
> Pythonで正規表現を使用して文字列からパターンを探す場合は、`re`モジュールを使用する。

## 1) Raw String

`r"..."`で表すRaw String (raw文字列) とは、文字列の先頭に `r` を付けることで、バックスラッシュ（`\`）をエスケープ文字として解釈せず、そのまま扱う文字列の表記方法である。


### ※ Raw Stringを使用する理由

- 正規表現の文法では、バックスラッシュ（`\`）を頻繁に使用する。

- しかし、通常の文字列では、そのたびにバックスラッシュを `\\` と記述する必要があり不便なため、Raw Stringを使用することが多い。

#### • 通常の文字列

通常の文字列では、`\n`が改行として解釈され、`\\`は1つのバックスラッシュとして出力される。

````python
print("Hello\nworld\\")
````

#### • Raw String

Raw Stringでは、`\n`や`\\`をエスケープシーケンスとして解釈せず、入力した形のまま文字列として扱われる。

````python
print(r"Hello\nworld\\")
````

## 2）文字クラス `[]`

文字クラス `[]` は、角括弧の中に指定した文字のうち、**いずれか1文字に一致するパターン**である。

````python
import re

text = "apple banana cat"

result = re.findall(r"[abc]", text)

print(result)
````

`[abc]` は、文字列の先頭から末尾までを調べ、`a`、`b`、`c` のいずれかに一致した文字を順番に返す。

## 3）文字クラスの範囲指定

文字クラス `[]` の中でハイフン（`-`）を使用すると、文字の範囲を表すことができる。


- 数字の0から9まで [0123456789] を省略して [0-9] と表記

- [a-z], [A-Z], ...

````python
import re

text = "A12B34"
result = re.findall(r"[0-9]", text)

print(result)

# 0から9までの数字のいずれか1文字に一致すると、返されるリストに追加される
````

#### - 英大文字を検索する

````python
import re

text = "Python JAVA AI"

result = re.findall(r"[A-Z]", text)

print(result)
````

## 4）文字の除外 `[^ ]`（否定文字クラス）

#### - 角括弧 `[]` の内側の先頭に `^`（caret）を置くと、角括弧内に指定した文字を除外し、それ以外の文字に一致する。

````python
import re

text = "a7bc123"

result = re.findall(r"[^0-9]", text) 

print(result)
````

#### - `^` を角括弧の外側でパターンの先頭に使用すると、文字列の先頭位置を表す。  

````python
import re

text = "7bc123"
result = re.findall(r"^[0-9]", text)

print(result)
````

````python
import re

password = input("パスワード：")
result = []

if re.search(r"[ぁ-んァ-ヶ一-龯々ー]", password):
    result.append("日本語は使用できません")

if not re.search(r"[a-z]", password):
    result.append("英小文字が含まれていません")

if not re.search(r"[A-Z]", password):
    result.append("英大文字が含まれていません")

if not re.search(r"[0-9]", password):
    result.append("数字が含まれていません")

if not re.search(r"[^0-9a-zA-Zぁ-んァ-ヶ一-龯々ー]", password):
    result.append("特殊文字が含まれていません")

if len(result) == 0:
    result.append("使用可能なパスワードです!")


result
````

## 5) 任意の1文字 `.`

`.` は、改行を除く任意の1文字を表す。

(文字の種類は問わないが、必ず**1文字だけ**に一致する)

````python
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # `c` で始まり、`t` で終わる3文字の文字列を検索する

print(result)
````

## 6) 文字列の先頭 `^` と末尾 `$`

- 文字列の途中からパターンを探すのではなく、文字列が特定の文字列で`始まるか、終わるか`を確認するときに使用する。

- `^` が `[]` の外にある場合  
  → `文字列の先頭`を表す。

- `^` が `[]` の先頭にある場合  
  → 角括弧内の文字`以外（否定）`を表す。

- `$`  
  → `文字列の末尾`を表す。

````python
import re

text = "Python Python python"

print(re.findall(r"^Py", text))
````

````python
text = "python Python python"

print(re.findall(r"^Py", text))
````

````python
texts = ["Python Python python", "python Python python"]

for t in texts:
    print(re.findall(r"^Py", t))
````

#### - `re.search(正規表現, text)` 関数を使用する。

#### - パターンに一致する部分が見つかった場合は `Match` オブジェクトを返し、見つからなかった場合は `None` を返す。

````python
import re

text = "Python Python python"

print(re.search(r"^Py", text))
print(re.search(r"^py", text))
````

````python
import re

text = "Python"

print(re.search(r"hon$", text))
print(re.search(r"abc$", text))
````

## 7) 量指定子 `*` と `+`

- `*`：直前の文字またはパターンが、`0回以上`繰り返されることを表す。  

- `+`：直前の文字またはパターンが、`1回以上`繰り返されることを表す。 

````python
import re

text = "Pyyyyython"

# Pyで始まる文字列を検索
print(re.search(r"^Py", text))

# Pで始まり、その後にyが1回以上続く文字列を検索
print(re.search(r"^Py*", text))
````

#### ※ * vs +

- `re.match()` 関数：文字列の先頭から、パターンに一致するかを確認する。

- `re.fullmatch()` 関数：文字列全体がパターンと完全に一致するかを確認する。

````python
import re

texts = ["123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d+：数字が1個以上必要である。
````

````python
import re

texts = ["123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d*", t))) # \d*：数字が0個以上である。空文字列もTrue。
````

## 8) 量指定子 `?`

- 直前の文字が、`0個または1個`存在することを表す。

- `ab?c`：`b` がない、または1個だけある場合に一致する。

  - `abc` → 一致する
  - `ac` → 一致する
  - `abbc` → `b` が2個あるため一致しない

````python
import re

words = [
    "ae",
    "abc",
    "abbcd",
    "abbbcde",
    "acde"
]

for word in words:
    print(re.match(r"ab?c", word))
````

## 9) バックスラッシュを使用する特殊文字

| パターン | 意味 | 例 |
|---|---|---|
| `\d` | 数字（`0～9`） | `123` |
| `\D` | 数字以外の文字 | `ABC` |
| `\w` | 英数字またはアンダースコア `_` | `abc_123` |
| `\W` | `\w` 以外の文字 | `!@#` |
| `\s` | 空白文字（スペース、タブ、改行など） | `スペース、タブ` |
| `\S` | 空白文字以外の文字 | `Python` |

````python
import re 

text = "Py@thon 123_xyz"

print(re.findall(r"\d", text))
print(re.findall(r"\D", text))
print(re.findall(r"\w", text))
print(re.findall(r"\W", text))
print(re.findall(r"\s", text))
print(re.findall(r"\S", text))
````

## 10) 量指定子 `{n}`

- 直前の文字またはパターンが、`ちょうどn回`繰り返されることを表す。

````python
import re

text = "1,23,456,7890,77a8888"

print(re.findall(r"\d{3}", text)) # 数字が3つ連続する部分だけを、結果のリストに追加する
````

````python
import re

text = "a bc def ghij kk@!#xx"

print(re.findall(r"\W{3}", text)) # 特殊文字が3つ連続する部分を検索する
````

## 11) 量指定子 `{n, m}`

- 直前の文字またはパターンが、`最小n回、最大m回`繰り返されることを表す。

````python
import re

text = "1 12 123 1234 12345"

print(re.findall(r"\d{2,4}", text))
````

## 12) OR演算子 `|`

- 複数のパターンのうち、`いずれか1つに一致`すればよいことを表す。

````python
import re

text = "cat, dog, bird, cats, dog"

print(re.findall(r"cat|dog", text))
````

## 13) `group()` メソッド

- Match オブジェクトから、実際に一致した文字列を取得する。

````python
import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)

    if result != None:
        print(result)
        print(result.group())
````
